```python
import os
import argparse
import warnings
import logging
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
from prophet import Prophet
from datetime import timedelta
 
warnings.filterwarnings('ignore')
logging.getLogger('prophet').setLevel(logging.ERROR)
logging.getLogger('cmdstanpy').setLevel(logging.ERROR)
 
# ─────────────────────────────────────────────
#  CONFIG  (edit these if needed)
# ─────────────────────────────────────────────
DATE_FORMAT   = "%d-%m-%Y"       # Change if your date format differs
FEATURES      = ['Temp_C', 'T_max_C', 'T_min_C',
                 'Humidity_pct', 'Wind_Speed_kmh', 'Solar_Rad_Wm2']
SEQ_LEN       = 90               # Days of history fed into GRU (lookback window)
HIDDEN_SIZE   = 128              # GRU hidden units
NUM_LAYERS    = 2                # GRU stacked layers
DROPOUT       = 0.2
BATCH_SIZE    = 64
EPOCHS        = 80
LR            = 1e-3
WEIGHT_DECAY  = 1e-5
PATIENCE      = 10               # Early stopping patience
VAL_SPLIT     = 0.15             # Fraction of data for validation
GRU_WEIGHT    = 0.55             # Weight of GRU in ensemble (Prophet gets 1 - this)
PROPHET_WEIGHT= 0.45
SEED          = 42
OUTPUT_DIR    = "forecast_outputs"
print('CELL 1 DONE')
```

```python
# ─────────────────────────────────────────────────────────────────────────────
#  1.  DATA LOADING & PREPROCESSING
# ─────────────────────────────────────────────────────────────────────────────

def load_data(csv_path: str) -> pd.DataFrame:
    """Load CSV, parse dates, sort, validate columns."""
    print(f"\n[1/6] Loading data from: {csv_path}")
    df = pd.read_csv(csv_path)

    # Try to find the date column (case-insensitive)
    date_col = next((c for c in df.columns if c.lower() == 'date'), None)
    if date_col is None:
        raise ValueError("No 'Date' column found in CSV.")
    df[date_col] = pd.to_datetime(df[date_col], format=DATE_FORMAT, dayfirst=True)
    df = df.rename(columns={date_col: 'Date'})
    df = df.sort_values('Date').reset_index(drop=True)

    missing = [f for f in FEATURES if f not in df.columns]
    if missing:
        raise ValueError(f"Missing columns: {missing}\nFound: {list(df.columns)}")

    df = df.dropna(subset=FEATURES)
    print(f"    Rows: {len(df):,}  |  Date range: {df['Date'].min().date()} → {df['Date'].max().date()}")
    return df


def scale_data(df: pd.DataFrame):
    """MinMax scale features. Returns scaled array and fitted scaler."""
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled = scaler.fit_transform(df[FEATURES].values.astype(np.float32))
    return scaled, scaler
```

```python
# ─────────────────────────────────────────────────────────────────────────────
#  2.  PYTORCH DATASET & GRU MODEL
# ─────────────────────────────────────────────────────────────────────────────

class WeatherDataset(Dataset):
    def __init__(self, data: np.ndarray, seq_len: int, horizon: int):
        self.X, self.y = [], []
        for i in range(len(data) - seq_len - horizon + 1):
            self.X.append(data[i : i + seq_len])
            self.y.append(data[i + seq_len : i + seq_len + horizon])
        self.X = torch.tensor(np.array(self.X), dtype=torch.float32)
        self.y = torch.tensor(np.array(self.y), dtype=torch.float32)

    def __len__(self):  return len(self.X)
    def __getitem__(self, idx): return self.X[idx], self.y[idx]


class GRUForecast(nn.Module):
    """
    Stacked GRU with:
      - Bidirectional first layer for richer context
      - Layer norm for training stability
      - Residual projection
      - Multi-step multi-variate output head
    """
    def __init__(self, input_size: int, hidden_size: int, num_layers: int,
                 horizon: int, n_features: int, dropout: float = 0.2):
        super().__init__()
        self.horizon    = horizon
        self.n_features = n_features

        self.gru = nn.GRU(
            input_size, hidden_size, num_layers,
            batch_first=True, dropout=dropout if num_layers > 1 else 0.0
        )
        self.norm   = nn.LayerNorm(hidden_size)
        self.drop   = nn.Dropout(dropout)
        self.fc1    = nn.Linear(hidden_size, hidden_size)
        self.act    = nn.GELU()
        self.fc_out = nn.Linear(hidden_size, horizon * n_features)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        out, _ = self.gru(x)
        out    = self.norm(out[:, -1, :])          # take last timestep
        out    = self.drop(self.act(self.fc1(out)))
        out    = self.fc_out(out)
        return out.view(-1, self.horizon, self.n_features)
```

```python
# ─────────────────────────────────────────────────────────────────────────────
#  3.  TRAINING LOOP
# ─────────────────────────────────────────────────────────────────────────────

def train_gru(scaled: np.ndarray, horizon: int, output_dir: str):
    """Train GRU with early stopping. Returns trained model + loss history."""
    print(f"\n[2/6] Training GRU model  (epochs={EPOCHS}, seq_len={SEQ_LEN}, horizon={horizon} days)")

    torch.manual_seed(SEED)
    np.random.seed(SEED)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"    Device: {device}")

    # Train/val split (time-aware — no shuffling across the split point)
    n          = len(scaled)
    val_start  = int(n * (1 - VAL_SPLIT))
    train_data = scaled[:val_start]
    val_data   = scaled[val_start - SEQ_LEN:]  # overlap for sequence building

    train_ds = WeatherDataset(train_data, SEQ_LEN, horizon)
    val_ds   = WeatherDataset(val_data,   SEQ_LEN, horizon)

    train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True,  num_workers=0)
    val_loader   = DataLoader(val_ds,   batch_size=BATCH_SIZE, shuffle=False, num_workers=0)

    model     = GRUForecast(len(FEATURES), HIDDEN_SIZE, NUM_LAYERS, horizon, len(FEATURES), DROPOUT).to(device)
    optimizer = torch.optim.AdamW(model.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)
    scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=EPOCHS, eta_min=LR * 0.01)
    criterion = nn.HuberLoss()   # robust to outliers vs plain MSE

    best_val_loss = float('inf')
    patience_ctr  = 0
    train_losses, val_losses = [], []
    best_state = None

    for epoch in range(1, EPOCHS + 1):
        # — Train —
        model.train()
        t_loss = 0.0
        for xb, yb in train_loader:
            xb, yb = xb.to(device), yb.to(device)
            optimizer.zero_grad()
            pred = model(xb)
            loss = criterion(pred, yb)
            loss.backward()
            nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            optimizer.step()
            t_loss += loss.item()
        t_loss /= len(train_loader)

        # — Validate —
        model.eval()
        v_loss = 0.0
        with torch.no_grad():
            for xb, yb in val_loader:
                xb, yb = xb.to(device), yb.to(device)
                v_loss += criterion(model(xb), yb).item()
        v_loss /= len(val_loader)

        scheduler.step()
        train_losses.append(t_loss)
        val_losses.append(v_loss)

        if epoch % 10 == 0 or epoch == 1:
            print(f"    Epoch {epoch:3d}/{EPOCHS}  |  train: {t_loss:.4f}  val: {v_loss:.4f}  lr: {scheduler.get_last_lr()[0]:.2e}")

        # — Early stopping —
        if v_loss < best_val_loss - 1e-5:
            best_val_loss = v_loss
            best_state    = {k: v.clone() for k, v in model.state_dict().items()}
            patience_ctr  = 0
        else:
            patience_ctr += 1
            if patience_ctr >= PATIENCE:
                print(f"    Early stopping at epoch {epoch}  (best val loss: {best_val_loss:.4f})")
                break

    # Restore best weights
    if best_state:
        model.load_state_dict(best_state)

    # Save model
    model_path = os.path.join(output_dir, "gru_model.pt")
    torch.save({"model_state": best_state, "config": {
        "input_size": len(FEATURES), "hidden_size": HIDDEN_SIZE,
        "num_layers": NUM_LAYERS, "horizon": horizon,
        "n_features": len(FEATURES), "dropout": DROPOUT,
        "seq_len": SEQ_LEN, "features": FEATURES
    }}, model_path)
    print(f"    Model saved → {model_path}")

    # Plot training curve
    _plot_training(train_losses, val_losses, output_dir)

    return model, device, train_losses, val_losses


def _plot_training(train_losses, val_losses, output_dir):
    fig, ax = plt.subplots(figsize=(8, 4))
    ax.plot(train_losses, label='Train Loss', color='#e74c3c')
    ax.plot(val_losses,   label='Val Loss',   color='#3498db')
    ax.set_xlabel('Epoch'); ax.set_ylabel('Huber Loss')
    ax.set_title('GRU Training Curve')
    ax.legend(); ax.grid(alpha=0.3)
    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, "training_loss.png"), dpi=120, bbox_inches='tight')
    plt.close()
    print(f"    Training curve saved → {output_dir}/training_loss.png")
```

```python
# ─────────────────────────────────────────────────────────────────────────────
#  4.  GRU INFERENCE
# ─────────────────────────────────────────────────────────────────────────────

def gru_predict(model, scaled: np.ndarray, scaler: MinMaxScaler,
                horizon: int, device) -> np.ndarray:
    """Run GRU on last SEQ_LEN days → forecast next `horizon` days."""
    model.eval()
    x = torch.tensor(scaled[-SEQ_LEN:], dtype=torch.float32).unsqueeze(0).to(device)
    with torch.no_grad():
        pred_scaled = model(x).squeeze(0).cpu().numpy()   # (horizon, n_features)
    return scaler.inverse_transform(pred_scaled)           # back to real units
```

```python
# ─────────────────────────────────────────────────────────────────────────────
#  5.  PROPHET FORECASTS
# ─────────────────────────────────────────────────────────────────────────────

def prophet_predict(df: pd.DataFrame, horizon: int) -> dict:
    """
    Fit one Prophet model per feature. Returns dict of
    {feature: DataFrame with ds, yhat, yhat_lower, yhat_upper}.
    """
    print(f"\n[3/6] Fitting Prophet models  ({len(FEATURES)} variables) ...")
    results = {}
    for feat in FEATURES:
        prophet_df = df[['Date', feat]].rename(columns={'Date': 'ds', feat: 'y'})
        m = Prophet(
            yearly_seasonality=True,
            weekly_seasonality=False,
            daily_seasonality=False,
            seasonality_mode='additive',
            changepoint_prior_scale=0.05,
            uncertainty_samples=500
        )
        m.fit(prophet_df)
        future   = m.make_future_dataframe(periods=horizon)
        forecast = m.predict(future)
        results[feat] = forecast[forecast['ds'] > df['Date'].max()][
            ['ds', 'yhat', 'yhat_lower', 'yhat_upper']
        ].reset_index(drop=True)
    print("    Prophet fitting complete.")
    return results
```

```python

# ─────────────────────────────────────────────────────────────────────────────
#  6.  ENSEMBLE & OUTPUT
# ─────────────────────────────────────────────────────────────────────────────

def build_ensemble(gru_pred: np.ndarray, prophet_results: dict,
                   last_date: pd.Timestamp, horizon: int) -> pd.DataFrame:
    """
    Weighted average: ensemble = GRU_WEIGHT * GRU + PROPHET_WEIGHT * Prophet.
    Confidence intervals come from Prophet (GRU is deterministic).
    """
    print(f"\n[4/6] Building ensemble  (GRU={GRU_WEIGHT:.0%}, Prophet={PROPHET_WEIGHT:.0%})")

    future_dates = [last_date + timedelta(days=i+1) for i in range(horizon)]
    records = []

    for i, date in enumerate(future_dates):
        row = {'Date': date}
        for j, feat in enumerate(FEATURES):
            gru_val     = gru_pred[i, j]
            p_row       = prophet_results[feat].iloc[i]
            prophet_val = p_row['yhat']
            lower       = p_row['yhat_lower']
            upper       = p_row['yhat_upper']

            ensemble    = GRU_WEIGHT * gru_val + PROPHET_WEIGHT * prophet_val
            # Shift CI proportionally around ensemble value
            ci_half     = (upper - lower) / 2
            row[feat]         = round(ensemble, 3)
            row[f'{feat}_low']  = round(ensemble - ci_half, 3)
            row[f'{feat}_high'] = round(ensemble + ci_half, 3)
        records.append(row)

    return pd.DataFrame(records)


def evaluate_on_holdout(df: pd.DataFrame, gru_pred_holdout: np.ndarray,
                        prophet_results_holdout: dict, holdout_df: pd.DataFrame):
    """Print MAE/RMSE on a small holdout set for transparency."""
    print("\n[5/6] Validation metrics on holdout (last 120 days of history):")
    print(f"    {'Feature':<22} {'GRU MAE':>10} {'Prophet MAE':>13} {'Ensemble MAE':>14}")
    print("    " + "-" * 61)
    for j, feat in enumerate(FEATURES):
        actual   = holdout_df[feat].values
        gru_v    = gru_pred_holdout[:, j]
        prop_v   = prophet_results_holdout[feat]['yhat'].values[:len(actual)]
        ens_v    = GRU_WEIGHT * gru_v + PROPHET_WEIGHT * prop_v
        g_mae    = mean_absolute_error(actual, gru_v)
        p_mae    = mean_absolute_error(actual, prop_v)
        e_mae    = mean_absolute_error(actual, ens_v)
        print(f"    {feat:<22} {g_mae:>10.3f} {p_mae:>13.3f} {e_mae:>14.3f}")


def save_outputs(forecast_df: pd.DataFrame, df: pd.DataFrame,
                 horizon: int, output_dir: str):
    """Save daily CSV, monthly summary CSV, and forecast plots."""
    print(f"\n[6/6] Saving outputs to ./{output_dir}/")
    os.makedirs(output_dir, exist_ok=True)

    # Trim to requested horizon (drop any extra Prophet days)
    last_history = df['Date'].max()
    forecast_df  = forecast_df[forecast_df['Date'] > last_history].head(horizon).copy()

    # ── Daily CSV ──
    daily_path = os.path.join(output_dir, "forecast_daily.csv")
    forecast_df.to_csv(daily_path, index=False)
    print(f"    Daily forecast     → {daily_path}  ({len(forecast_df)} rows)")

    # ── Monthly summary ──
    forecast_df['Month'] = forecast_df['Date'].dt.to_period('M')
    monthly = forecast_df.groupby('Month')[FEATURES].mean().round(2)
    monthly_path = os.path.join(output_dir, "forecast_monthly_summary.csv")
    monthly.to_csv(monthly_path)
    print(f"    Monthly summary    → {monthly_path}")
    print("\n    ── Monthly Forecast Summary ──")
    print(monthly.to_string())

    # ── Plots ──
    _plot_forecast(forecast_df, df, output_dir)

    return forecast_df, monthly


def _plot_forecast(forecast_df: pd.DataFrame, hist_df: pd.DataFrame, output_dir: str):
    colors = ['#e74c3c','#f39c12','#3498db','#27ae60','#9b59b6','#e67e22']
    labels = ['Avg Temp (°C)','Max Temp (°C)','Min Temp (°C)',
              'Humidity (%)','Wind Speed (km/h)','Solar Rad (W/m²)']

    fig = plt.figure(figsize=(18, 14))
    fig.suptitle('4-Month Weather Forecast  —  GRU + Prophet Ensemble\n'
                 f'{forecast_df["Date"].min().strftime("%b %d, %Y")} → '
                 f'{forecast_df["Date"].max().strftime("%b %d, %Y")}',
                 fontsize=15, fontweight='bold', y=0.99)

    gs = gridspec.GridSpec(3, 2, figure=fig, hspace=0.45, wspace=0.3)

    hist_window = hist_df[hist_df['Date'] >= hist_df['Date'].max() - pd.Timedelta(days=180)]

    for idx, (feat, color, label) in enumerate(zip(FEATURES, colors, labels)):
        ax = fig.add_subplot(gs[idx // 2, idx % 2])

        # Historical
        ax.plot(hist_window['Date'], hist_window[feat],
                color='#7f8c8d', alpha=0.6, linewidth=0.9, label='Historical')

        # Confidence band
        ax.fill_between(forecast_df['Date'],
                        forecast_df[f'{feat}_low'],
                        forecast_df[f'{feat}_high'],
                        alpha=0.18, color=color)

        # Forecast line
        ax.plot(forecast_df['Date'], forecast_df[feat],
                color=color, linewidth=2.2, label='Ensemble Forecast')

        # Divider
        ax.axvline(hist_df['Date'].max(), color='black',
                   linestyle='--', linewidth=0.9, alpha=0.5)

        ax.set_title(label, fontsize=11, fontweight='bold', pad=6)
        ax.tick_params(axis='x', rotation=30, labelsize=7.5)
        ax.tick_params(axis='y', labelsize=8.5)
        ax.grid(alpha=0.25, linestyle=':')
        ax.legend(fontsize=7.5, loc='best')

    plot_path = os.path.join(output_dir, "forecast_plots.png")
    plt.savefig(plot_path, dpi=150, bbox_inches='tight')
    plt.close()
    print(f"    Forecast plots     → {plot_path}")
```

```python
# ─────────────────────────────────────────────────────────────────────────────
#  MAIN
# ─────────────────────────────────────────────────────────────────────────────

# --- SET RUN VARIABLES HERE ---
CSV_FILE_PATH = "refined_weather_2012-2025.csv" 
FORECAST_MONTHS = 4
FAST_MODE = False # Set to True for a quick 20-epoch CPU test run

if FAST_MODE:
    EPOCHS = 20
    HIDDEN_SIZE = 64
    BATCH_SIZE = 128
    print("  [--fast mode] Reduced epochs and hidden size for quick run.")

horizon = FORECAST_MONTHS * 30
os.makedirs(OUTPUT_DIR, exist_ok=True)

print("=" * 70)
print(f"  WEATHER FORECAST MODEL  |  Horizon: {FORECAST_MONTHS} months ({horizon} days)")
print("=" * 70)

# Load & Scale
df = load_data(CSV_FILE_PATH)
scaled, scaler = scale_data(df)

# Holdout prep
holdout_df   = df.iloc[-horizon:].copy()
train_df     = df.iloc[:-horizon].copy()
scaled_train = scaled[:-horizon]

# Train GRU
model, device, _, _ = train_gru(scaled_train, horizon, OUTPUT_DIR)

# GRU Inference
scaled_for_holdout = scaled[:-(horizon)]
gru_holdout = gru_predict(model, scaled_for_holdout, scaler, horizon, device)
gru_future  = gru_predict(model, scaled, scaler, horizon, device)

# Prophet on train (for holdout metrics)
print("\n    Fitting Prophet on training split (for metrics)...")
prophet_holdout = {}
for feat in FEATURES:
    pdf = train_df[['Date', feat]].rename(columns={'Date': 'ds', feat: 'y'})
    m = Prophet(yearly_seasonality=True, weekly_seasonality=False,
                daily_seasonality=False, seasonality_mode='additive',
                changepoint_prior_scale=0.05, uncertainty_samples=200)
    m.fit(pdf)
    future_pdf = m.make_future_dataframe(periods=horizon)
    fc = m.predict(future_pdf)
    prophet_holdout[feat] = fc[fc['ds'] > train_df['Date'].max()][['ds','yhat','yhat_lower','yhat_upper']].reset_index(drop=True)

# Metrics
evaluate_on_holdout(df, gru_holdout, prophet_holdout, holdout_df)

# Prophet full prediction
prophet_future = prophet_predict(df, horizon)

# Build Ensemble & Save
forecast_df = build_ensemble(gru_future, prophet_future, df['Date'].max(), horizon)
save_outputs(forecast_df, df, horizon, OUTPUT_DIR)

print("\n" + "=" * 70)
print("  DONE. All outputs saved to:", os.path.abspath(OUTPUT_DIR))
print("=" * 70 + "\n")
```

```python
'''===============================================================================
  evaluate_holdout_metrics.py
  ─────────────────────────────────────────────────────────────────────────────
  DROP-IN evaluation cell for weatherforecast.ipynb.
 
  Computes four paper-required metrics on the 120-day holdout set and prints
  a side-by-side comparison table: GRU Only vs. Ensemble.
 
  Metrics
  -------
    MAE      – Mean Absolute Error
    RMSE     – Root Mean Square Error
    R²       – Coefficient of Determination
    Accuracy – Regression accuracy approximation:
               100 × (1 − MAE / |mean(y_true)|)
               Clipped to [0, 100] so negative scores read as 0 %.
 
  Expected notebook globals (already defined in your cells)
  ---------------------------------------------------------
    holdout_df               : pd.DataFrame  – ground-truth (120 rows)
    gru_pred_holdout         : np.ndarray    – shape (120, n_features)
    prophet_results_holdout  : dict          – {feat: DataFrame w/ 'yhat'}
    FEATURES                 : list[str]
    GRU_WEIGHT, PROPHET_WEIGHT : float
 
  Usage
  -----
    Paste this cell anywhere after the variables above are in scope, then run:
        results = evaluate_holdout_extended(
            holdout_df,
            gru_pred_holdout,
            prophet_results_holdout,
            FEATURES,
            GRU_WEIGHT,
            PROPHET_WEIGHT,
        )
================================================================================
'''
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
 
 
# ─────────────────────────────────────────────────────────────────────────────
#  Helper: compute all four metrics for one (y_true, y_pred) pair
# ─────────────────────────────────────────────────────────────────────────────
 
def _compute_metrics(y_true: np.ndarray, y_pred: np.ndarray) -> dict:
    """
    Return MAE, RMSE, R², and Accuracy for a single feature's arrays.
 
    Parameters
    ----------
    y_true : 1-D numpy array of ground-truth values
    y_pred : 1-D numpy array of predicted values (same length)
 
    Returns
    -------
    dict with keys: 'MAE', 'RMSE', 'R2', 'Accuracy'
    """
    mae  = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2   = r2_score(y_true, y_pred)
 
    true_mean = np.mean(np.abs(y_true))
    if true_mean == 0:
        # Edge case: constant zero series → accuracy undefined, set to 0
        accuracy = 0.0
    else:
        accuracy = max(0.0, 100.0 * (1.0 - mae / true_mean))
 
    return {"MAE": mae, "RMSE": rmse, "R2": r2, "Accuracy": accuracy}
 
```

```python
# ─────────────────────────────────────────────────────────────────────────────
#  Main evaluation function
# ─────────────────────────────────────────────────────────────────────────────
 
def evaluate_holdout_extended(
    holdout_df: "pd.DataFrame",
    gru_pred_holdout: np.ndarray,
    prophet_results_holdout: dict,
    features: list,
    gru_weight: float,
    prophet_weight: float,
) -> dict:
    """
    Evaluate GRU-only and Ensemble predictions on the holdout set.
 
    Prints a formatted side-by-side comparison table with four metrics
    per feature, then returns the raw results for downstream use.
 
    Parameters
    ----------
    holdout_df              : DataFrame with ground-truth columns matching `features`
    gru_pred_holdout        : ndarray, shape (n_days, n_features) – GRU predictions
    prophet_results_holdout : dict mapping feature name → DataFrame with 'yhat' column
    features                : list of feature column name strings (e.g. FEATURES)
    gru_weight              : float blend weight for GRU (e.g. GRU_WEIGHT = 0.55)
    prophet_weight          : float blend weight for Prophet (e.g. PROPHET_WEIGHT = 0.45)
 
    Returns
    -------
    dict  →  { feature: { 'GRU': {MAE, RMSE, R2, Accuracy},
                           'Ensemble': {MAE, RMSE, R2, Accuracy} } }
    """
 
    n_days   = len(holdout_df)
    results  = {}
 
    # ── Column widths ────────────────────────────────────────────────────────
    feat_w  = max(len(f) for f in features) + 2   # feature name column
    col_w   = 10                                   # metric value column
    model_w = col_w * 4 + 3                        # 4 metrics + separators
 
    metric_labels = ["MAE", "RMSE", "R²", "Accuracy%"]
 
    # ── Header ───────────────────────────────────────────────────────────────
    divider = "─" * (feat_w + 2 + model_w * 2 + 7)
 
    print()
    print("=" * len(divider))
    print(
        f"  HOLDOUT EVALUATION  |  {n_days}-day set  |  "
        f"Ensemble blend: GRU {gru_weight*100:.0f}% / Prophet {prophet_weight*100:.0f}%"
    )
    print("=" * len(divider))
 
    # Sub-header: model group labels
    gru_header = "GRU Only".center(model_w)
    ens_header = "Ensemble (GRU + Prophet)".center(model_w)
    print(f"  {'Feature':<{feat_w}}  {gru_header}  {ens_header}")
 
    # Sub-header: metric labels (repeated per model group)
    metric_hdr = "  ".join(f"{m:>{col_w}}" for m in metric_labels)
    print(f"  {'':^{feat_w}}  {metric_hdr}  {metric_hdr}")
    print(f"  {divider}")
 
    # ── Per-feature rows ─────────────────────────────────────────────────────
    for j, feat in enumerate(features):
        y_true = holdout_df[feat].values.astype(float)
 
        # GRU predictions for this feature
        gru_pred = gru_pred_holdout[:, j].astype(float)
 
        # Prophet predictions — align length to holdout
        prophet_pred = (
            prophet_results_holdout[feat]["yhat"]
            .values[:n_days]
            .astype(float)
        )
 
        # Ensemble blend
        ensemble_pred = gru_weight * gru_pred + prophet_weight * prophet_pred
 
        # Compute metrics
        gru_m = _compute_metrics(y_true, gru_pred)
        ens_m = _compute_metrics(y_true, ensemble_pred)
 
        results[feat] = {"GRU": gru_m, "Ensemble": ens_m}
 
        # Format row values
        def _fmt(m: dict) -> str:
            return "  ".join([
                f"{m['MAE']:>{col_w}.4f}",
                f"{m['RMSE']:>{col_w}.4f}",
                f"{m['R2']:>{col_w}.4f}",
                f"{m['Accuracy']:>{col_w}.2f}",
            ])
 
        print(f"  {feat:<{feat_w}}  {_fmt(gru_m)}  {_fmt(ens_m)}")
 
    # ── Aggregate averages ────────────────────────────────────────────────────
    print(f"  {divider}")
 
    def _avg(model_key: str, metric_key: str) -> float:
        return np.mean([results[f][model_key][metric_key] for f in features])
 
    avg_gru = {
        "MAE":      _avg("GRU", "MAE"),
        "RMSE":     _avg("GRU", "RMSE"),
        "R2":       _avg("GRU", "R2"),
        "Accuracy": _avg("GRU", "Accuracy"),
    }
    avg_ens = {
        "MAE":      _avg("Ensemble", "MAE"),
        "RMSE":     _avg("Ensemble", "RMSE"),
        "R2":       _avg("Ensemble", "R2"),
        "Accuracy": _avg("Ensemble", "Accuracy"),
    }
 
    print(f"  {'AVERAGE':<{feat_w}}  {_fmt(avg_gru)}  {_fmt(avg_ens)}")
    print(f"  {'=' * len(divider)}")
    print()
 
    # ── Δ improvement row (positive = Ensemble is better) ────────────────────
    print("  Ensemble improvement over GRU  (↓ better for MAE/RMSE | ↑ better for R²/Accuracy%)")
    print(f"  {'─' * len(divider)}")
 
    def _delta_fmt(key_better_low, key_better_high):
        """Print delta row for each feature."""
        for feat in features:
            gm = results[feat]["GRU"]
            em = results[feat]["Ensemble"]
            d_mae  = gm["MAE"]      - em["MAE"]        # + means Ensemble is better
            d_rmse = gm["RMSE"]     - em["RMSE"]
            d_r2   = em["R2"]       - gm["R2"]
            d_acc  = em["Accuracy"] - gm["Accuracy"]
 
            def _sign(v):
                return f"+{v:.4f}" if v >= 0 else f"{v:.4f}"
 
            row = "  ".join([
                f"{_sign(d_mae):>{col_w}}",
                f"{_sign(d_rmse):>{col_w}}",
                f"{_sign(d_r2):>{col_w}}",
                f"{_sign(d_acc):>{col_w}}",
            ])
            print(f"  {feat:<{feat_w}}  {row}")
 
    _delta_fmt(None, None)   # args unused; kept for readability
    print(f"  {'─' * len(divider)}")
    print()
 
    return results
 
```

```python
# ─────────────────────────────────────────────────────────────────────────────
#  CALL  — uses the globals already defined in your notebook
# ─────────────────────────────────────────────────────────────────────────────
 
results = evaluate_holdout_extended(
    holdout_df=holdout_df,
    gru_pred_holdout=gru_pred_holdout,
    prophet_results_holdout=prophet_results_holdout,
    features=FEATURES,
    gru_weight=GRU_WEIGHT,
    prophet_weight=PROPHET_WEIGHT,
)
```
