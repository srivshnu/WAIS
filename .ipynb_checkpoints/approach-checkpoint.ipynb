# CDCIS Approach Notebook
### Cauvery Delta Crop Intelligence System — Pipeline Setup

**Stages covered here:**
1. Load data
2. Fetch & filter Thanjavur block boundaries (OSM)
3. Fetch canal/river network (OSM)
4. Pull satellite weather (Google Earth Engine — CHIRPS)
5. Elimination gate engine (G1–G7, FAO-56 compliant)

> Run cells top-to-bottom. Cells 2–4 require internet access.

## 1. Load Data

```python
# Cell 1 — Load all static input files
import pandas as pd

# BUG FIX 1: 'optimal values.csv' had a space in the filename. Use the canonical
#            underscore name from the product spec. 'refined weather 2012-2023.csv'
#            also had a space — fixed below.
# BUG FIX 2: All four files must exist in the working directory before running.

soil_data     = pd.read_csv("cleaned_thanjavur_full_dataset.csv")
lethal_limits = pd.read_csv("lethal_limits_for_crops.csv")
optimal_values = pd.read_csv("optimal_values.csv")          # was: 'optimal values.csv'
weather_data  = pd.read_csv("refined_weather_2012-2023.csv") # was: 'refined weather 2012-2023.csv'

print("Loaded soil_data     :", soil_data.shape)
print("Loaded lethal_limits :", lethal_limits.shape)
print("Loaded optimal_values:", optimal_values.shape)
print("Loaded weather_data  :", weather_data.shape)
print("\nStep 1 done. Data loaded.")
```

## 2. Fetch & Filter Thanjavur Block Boundaries (OSM)

```python
# Cell 2 — Fetch Thanjavur block-level admin boundaries from OpenStreetMap
# BUG FIX: OpenStreetMap's geocoder fails on the text "Thanjavur District". 
#          To guarantee stability, we bypass text search entirely and fetch
#          by GPS bounding box. Cell 3 will filter the excess neighboring blocks.

import osmnx as ox
import geopandas as gpd

print("Fetching regional taluk boundaries via GPS Bounding Box...")

# A GPS box that comfortably covers Thanjavur and its immediate surroundings
north, south, east, west = 11.35, 10.10, 79.70, 78.80
tags = {"admin_level": "8"}   # 8 = taluk/block in Tamil Nadu OSM schema

# Fetch raw map features within these coordinates
region_blocks = ox.features_from_bbox(north=north, south=south, east=east, west=west, tags=tags)

# Reset multi-index so 'name' becomes a regular column
region_blocks = region_blocks.reset_index()

# Keep only polygons (admin boundaries are always Polygon/MultiPolygon)
region_blocks = region_blocks[
    region_blocks.geometry.type.isin(["Polygon", "MultiPolygon"])
].copy()

# Retain only what we need
keep_cols = [c for c in ["name", "geometry"] if c in region_blocks.columns]
region_blocks = region_blocks[keep_cols]

region_blocks.to_file("thanjavur_blocks.geojson", driver="GeoJSON")

print(f"Fetched {len(region_blocks)} polygon features from the region.")
print("Saved → thanjavur_blocks.geojson")
```

```python
# Cell 3 — Filter down to the exact 14 Thanjavur District revenue blocks
# BUG FIX 5: Original list had 'Thiruvonam' — this is a star constellation /
#            festival name, NOT a Thanjavur block. The correct 14th block is
#            'Thirukattupalli'. Updated list matches official TNAU/district records.
# BUG FIX 6: The original code used json.load then iterated GeoJSON manually.
#            Using geopandas.read_file is safer — it handles CRS, encoding,
#            and geometry types correctly. str.contains with case=False is
#            more robust than manual lower() + 'in' checks.

import geopandas as gpd

THANJAVUR_14_BLOCKS = [
    "Thanjavur", "Kumbakonam", "Papanasam", "Pattukkottai",
    "Peravurani", "Orathanadu", "Thiruvidaimarudur", "Thiruvaiyaru",
    "Budalur", "Ammapettai", "Madukkur", "Sethubhavachatram",
    "Thiruppanandal", "Thirukattupalli"  # FIX: was 'Thiruvonam'
]

gdf = gpd.read_file("thanjavur_blocks.geojson")

# Build a case-insensitive regex pattern from the authoritative block list
import re
pattern = "|".join(re.escape(b) for b in THANJAVUR_14_BLOCKS)
mask = gdf["name"].str.contains(pattern, case=False, na=False, regex=True)
clean_gdf = gdf[mask].copy()

clean_gdf.to_file("thanjavur_14_blocks.geojson", driver="GeoJSON")

print(f"Filtered to {len(clean_gdf)} blocks:")
for name in sorted(clean_gdf["name"].tolist()):
    print(f"  ✓ {name}")

if len(clean_gdf) != 14:
    print(f"\n⚠️  WARNING: Expected 14 blocks, got {len(clean_gdf)}.")
    print("   Some OSM names may differ — check names above and update THANJAVUR_14_BLOCKS.")
```

## 3. Fetch Canal & River Network (OSM)

```python
# Cell 4 — Fetch Cauvery canal and river network within Thanjavur boundary
# BUG FIX 7: Original code called .unary_union on the GeoDataFrame directly.
#            unary_union is a property of the GeoSeries (gdf.geometry), not
#            the GeoDataFrame itself. Fix: gdf.geometry.unary_union.
# BUG FIX 8: ox.features_from_polygon returns a MultiIndex GeoDataFrame.
#            Must call .reset_index() before column selection or 'waterway'
#            and 'name' won't exist as regular columns — causes KeyError.
# BUG FIX 9: cols_to_keep check used 'name' in canals_gdf.columns but after
#            reset_index() some OSM tag columns are missing entirely.
#            Use .get() with fallback to avoid silent column drops.

import osmnx as ox
import geopandas as gpd

print("Reading 14-block GeoJSON...")
blocks_gdf = gpd.read_file("thanjavur_14_blocks.geojson")

# FIX 7: .geometry.unary_union (not .unary_union on the GeoDataFrame)
thanjavur_boundary = blocks_gdf.geometry.unary_union

print("Fetching waterways from OpenStreetMap... (may take 30-60 sec)")
tags = {"waterway": ["canal", "river"]}
canals_gdf = ox.features_from_polygon(thanjavur_boundary, tags)

# FIX 8: reset MultiIndex before column access
canals_gdf = canals_gdf.reset_index()

# Keep only line geometries (rivers/canals are linestrings in OSM)
canals_gdf = canals_gdf[
    canals_gdf.geometry.type.isin(["LineString", "MultiLineString", "Polygon"])
].copy()

# FIX 9: safe column selection
cols_to_keep = ["geometry"]
for col in ["waterway", "name"]:
    if col in canals_gdf.columns:
        cols_to_keep.append(col)

canals_clean = canals_gdf[cols_to_keep].reset_index(drop=True)
canals_clean.to_file("thanjavur_canals.geojson", driver="GeoJSON")

print(f"Saved {len(canals_clean)} waterway segments → thanjavur_canals.geojson")
if "name" in canals_clean.columns:
    major = canals_clean["name"].dropna().unique()[:10]
    print("Major waterways (sample):", list(major))
```

## 4. Pull Satellite Weather Data (Google Earth Engine — CHIRPS)

```python
# Cell 5 — Extract daily CHIRPS rainfall per block via Google Earth Engine
#
# BUG FIX 10: The original cell had a massive Windows/Python3 compatibility
#             hack (mocking fcntl, termios, StringIO, cStringIO). These are
#             NOT needed on Linux or macOS, and on Windows the correct fix is
#             to install the earthengine-api package properly — NOT to mock
#             standard library modules, which can corrupt other imports.
#             Removed entirely.
#
# BUG FIX 11: ee.Initialize() with no arguments fails if no credentials
#             are cached. The correct modern pattern (earthengine-api >= 0.1.370)
#             is ee.Initialize(project='your-gcp-project'). Added project param
#             with a clear instruction placeholder.
#
# BUG FIX 12: The loop variable 'f' shadowed the built-in open() function
#             (used earlier in the same scope as `with open('...') as f`).
#             Renamed loop variable to 'feat'.
#
# BUG FIX 13: props.get('mean', 0.0) silently converts missing data to 0.0 —
#             which is physically incorrect (0mm rain ≠ no data). Use None
#             and let the DataFrame preserve NaN for missing values.
#
# BUG FIX 14: The extractor mapped the reducer output to 'mean' but CHIRPS
#             band is named 'precipitation'. Must select the band explicitly
#             before reduceRegions, or rename after. Fixed with .select().
#
# ⚠️  REQUIREMENT: Set GCP_PROJECT below before running.
#    Run `earthengine authenticate` in terminal first if not yet authenticated.

import ee
import json
import pandas as pd

GCP_PROJECT = "your-gcp-project-id"  # <-- REPLACE with your GCP project ID

try:
    ee.Initialize(project=GCP_PROJECT)
    print("✅ Earth Engine Initialised!")
except Exception:
    print("No cached credentials found. Starting authentication flow...")
    ee.Authenticate()
    ee.Initialize(project=GCP_PROJECT)
    print("✅ Earth Engine Initialised!")

# Load block boundaries
with open("thanjavur_14_blocks.geojson", "r") as fh:
    geojson_data = json.load(fh)

features = []
for feat in geojson_data["features"]:          # FIX 12: was 'f', shadows open()
    geom = ee.Geometry(feat["geometry"])
    name = (feat["properties"].get("name")
            or feat["properties"].get("taluk", "Unknown"))
    features.append(ee.Feature(geom, {"block_name": name}))

thanjavur_fc = ee.FeatureCollection(features)

# --- Date range (edit as needed) ---
start_date = "2023-06-01"
end_date   = "2023-11-30"   # Full Kharif season
print(f"Pulling CHIRPS daily rainfall: {start_date} → {end_date}")

def extract_daily(image):
    date = image.date().format("YYYY-MM-dd")
    # FIX 14: select 'precipitation' band explicitly before reduceRegions
    stats = (image.select("precipitation")
                  .reduceRegions(
                      collection=thanjavur_fc,
                      reducer=ee.Reducer.mean(),
                      scale=5566
                  ))
    return stats.map(lambda f: f.set("date", date))

chirps = (ee.ImageCollection("UCSB-CHG/CHIRPS/DAILY")
            .filterDate(start_date, end_date))

daily_rain = chirps.map(extract_daily).flatten().getInfo()

# Build DataFrame — FIX 13: preserve NaN for genuinely missing pixels
records = []
for feat in daily_rain["features"]:
    props = feat["properties"]
    records.append({
        "Date":               props.get("date"),
        "Block":              props.get("block_name"),
        "Daily_Rainfall_mm":  round(props["mean"], 2) if props.get("mean") is not None else None
    })

df = pd.DataFrame(records).sort_values(["Block", "Date"]).reset_index(drop=True)
df.to_csv("thanjavur_satellite_weather.csv", index=False)

print(f"✅ Extraction complete: {len(df)} records → thanjavur_satellite_weather.csv")
print(df.head(10).to_string())
```

## 5. Elimination Gate Engine (G1–G7, FAO-56 Compliant)

```python
# Cell 6 — FAO-56 / CDCIS compliant elimination gate engine
#
# BUG FIX 15: G4 (Flowering Deficit) used water_sim_df['Stage'] == 'Mid'
#             to isolate the flowering window. This is wrong — the framework
#             spec uses ΣGDD against GDD_start/GDD_end of Stage_Critical==1
#             from scientific_crop_stages_fao56.csv. The stage name 'Mid' is
#             just a label; the critical flag is the authoritative source.
#             The daily sim df must carry a 'Stage_Critical' column, not 'Stage'.
#
# BUG FIX 16: G4 gate threshold in original = crop_row['Lethal_Flowering_Ks'].
#             Framework spec says: threshold = Lethal_Flowering_Ks × 0.95
#             (a 5% buffer to avoid hard boundary artifacts). Fixed.
#
# BUG FIX 17: G5 (Cumulative Drought) original formula was:
#               fdi = stress_days / total_days   (fraction of stressed days)
#             Framework spec (Formula B0-FDI) defines FDI as:
#               FDI = ΣDr_excess / (TAW_mean × N_season_days)
#             where Dr_excess_t = max(0, Dr_t - RAW_t). This is a volumetric
#             drought severity index, NOT just a count of stressed days.
#             Fixed to match spec. water_sim_df must carry 'Dr', 'RAW', 'TAW'.
#
# BUG FIX 18: G5 gate threshold in original = crop_row['Lethal_Drought_Index']
#             (column doesn't exist). Correct column from lethal_limits spec
#             is 'Lethal_Drought_FDI', and the gate threshold = value × 1.10
#             per framework. Fixed column name and multiplier.
#
# BUG FIX 19: Function had no docstring explaining expected columns in
#             water_sim_df. Added explicit contract comment.

import pandas as pd
import numpy as np

def evaluate_all_gates(crop_row: pd.Series,
                        max_temp: float,
                        total_rain: float,
                        soil_ph: float,
                        soil_n: float,
                        projected_gw_class: int,
                        water_sim_df: pd.DataFrame) -> tuple[bool, list[str]]:
    """
    Runs one crop through all 7 CDCIS elimination gates.

    Parameters
    ----------
    crop_row          : Row from lethal_limits_for_crops.csv
    max_temp          : Season peak temperature (°C)
    total_rain        : Season cumulative rainfall (mm)
    soil_ph           : Block soil pH
    soil_n            : Block soil nitrogen (kg/ha)
    projected_gw_class: Projected groundwater class (1–4, WSI-adjusted)
    water_sim_df      : Daily FAO-56 simulation DataFrame with columns:
                          - Stage_Critical (int, 1 = flowering window)
                          - Ks            (float, [0,1] water stress coeff)
                          - Dr            (float, root zone depletion mm)
                          - RAW           (float, readily available water mm)
                          - TAW           (float, total available water mm)

    Returns
    -------
    (survived: bool, elimination_reasons: list[str])
    """
    reasons = []

    # ── G1: Heat Sterility ───────────────────────────────────────────────────
    if max_temp >= crop_row["Lethal_Heat_C"]:
        reasons.append(
            f"G1: Peak temp ({max_temp}°C) ≥ sterility threshold "
            f"({crop_row['Lethal_Heat_C']}°C)"
        )

    # ── G2: Flood / Waterlogging ─────────────────────────────────────────────
    if total_rain >= crop_row["Lethal_Flood_mm"]:
        reasons.append(
            f"G2: Season rainfall ({total_rain}mm) ≥ flood limit "
            f"({crop_row['Lethal_Flood_mm']}mm)"
        )

    # ── G3: Groundwater Class ────────────────────────────────────────────────
    if projected_gw_class < crop_row["Lethal_GW_Class"]:
        reasons.append(
            f"G3: Projected GW class ({projected_gw_class}) < crop minimum "
            f"({crop_row['Lethal_GW_Class']})"
        )

    # ── G6: Soil pH ──────────────────────────────────────────────────────────
    if soil_ph < crop_row["Lethal_pH_min"]:
        reasons.append(
            f"G6a: Soil pH ({soil_ph}) < crop minimum ({crop_row['Lethal_pH_min']})"
        )
    elif soil_ph > crop_row["Lethal_pH_max"]:
        reasons.append(
            f"G6b: Soil pH ({soil_ph}) > crop maximum ({crop_row['Lethal_pH_max']})"
        )

    # ── G7: Soil Nitrogen ────────────────────────────────────────────────────
    if soil_n < crop_row["Lethal_N_min_kgha"]:
        reasons.append(
            f"G7: Soil N ({soil_n} kg/ha) < crop minimum "
            f"({crop_row['Lethal_N_min_kgha']} kg/ha)"
        )

    # ── Dynamic gates (require water simulation output) ──────────────────────
    if water_sim_df is not None and not water_sim_df.empty:

        # G4: Flowering Stage Deficit
        # FIX 15: use Stage_Critical flag, not stage name string
        # FIX 16: gate threshold = Lethal_Flowering_Ks × 0.95
        flowering = water_sim_df[water_sim_df["Stage_Critical"] == 1]
        if not flowering.empty:
            mean_ks = flowering["Ks"].mean()
            ks_threshold = crop_row["Lethal_Flowering_Ks"] * 0.95   # 5% buffer
            if mean_ks < ks_threshold:
                reasons.append(
                    f"G4: Mean Ks during flowering ({mean_ks:.3f}) < threshold "
                    f"({ks_threshold:.3f}, Lethal={crop_row['Lethal_Flowering_Ks']})"
                )

        # G5: Cumulative Drought — Farm Drought Index (FAO-56 B0-FDI)
        # FIX 17: volumetric FDI formula from framework spec
        # FIX 18: correct column name + × 1.10 gate multiplier
        sim = water_sim_df.copy()
        sim["Dr_excess"] = np.maximum(0, sim["Dr"] - sim["RAW"])
        sigma_dr_excess  = sim["Dr_excess"].sum()
        taw_mean         = sim["TAW"].mean()
        n_days           = len(sim)

        fdi = sigma_dr_excess / (taw_mean * n_days) if (taw_mean * n_days) > 0 else 0.0

        # FIX 18: column is 'Lethal_Drought_FDI', gate = value × 1.10
        fdi_gate = crop_row["Lethal_Drought_FDI"] * 1.10
        if fdi >= fdi_gate:
            reasons.append(
                f"G5: FDI ({fdi:.3f}) ≥ gate threshold ({fdi_gate:.3f}, "
                f"Lethal={crop_row['Lethal_Drought_FDI']})"
            )

    survived = len(reasons) == 0
    return survived, reasons


# ── Quick smoke test with synthetic data ────────────────────────────────────
print("Running smoke test on gate engine...")

dummy_crop = pd.Series({
    "Crop": "paddy",
    "Lethal_Heat_C": 38.0,
    "Lethal_Flood_mm": 1800.0,
    "Lethal_GW_Class": 2,
    "Lethal_pH_min": 5.0,
    "Lethal_pH_max": 8.0,
    "Lethal_N_min_kgha": 80.0,
    "Lethal_Flowering_Ks": 0.65,
    "Lethal_Drought_FDI": 0.40,
})

dummy_sim = pd.DataFrame({
    "Stage_Critical": [0]*30 + [1]*60 + [0]*30,  # Mid = critical
    "Ks":  [1.0]*80 + [0.45]*40,                  # stressed in late season
    "Dr":  [20.0]*80 + [80.0]*40,
    "RAW": [50.0]*120,
    "TAW": [100.0]*120,
})

survived, reasons = evaluate_all_gates(
    crop_row=dummy_crop,
    max_temp=36.0,
    total_rain=1200.0,
    soil_ph=6.5,
    soil_n=100.0,
    projected_gw_class=3,
    water_sim_df=dummy_sim,
)

print(f"Survived: {survived}")
if reasons:
    for r in reasons:
        print(f"  ✗ {r}")
else:
    print("  ✓ All gates passed")

print("\n✅ Gate engine ready.")
```
