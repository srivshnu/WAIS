### Fix A · `lethal_limits_for_crops.csv`
1. Drops the units header row (`—` in Crop column).
2. Renames `Lethal_Drought_Index` → `Lethal_Drought_FDI` (gate engine key).
3. Adds `season` column (Kharif / Rabi / All) from TNAU sowing calendar.

```python
# Fix lethal_limits_for_crops.csv — 3 issues:
# 1. Row 0 is a units header row (e.g. "°C (sterility)") — not data, must be dropped.
# 2. Column "Lethal_Drought_Index" must be renamed to "Lethal_Drought_FDI"
#    (gate engine uses Lethal_Drought_FDI; wrong name causes KeyError).
# 3. No "season" column — needed to filter crops by Kharif/Rabi/All.
#    Assigned from TNAU/FAO-56 seasonal sowing calendar for Cauvery Delta.

import pandas as pd

df = pd.read_csv("lethal_limits_for_crops.csv")

# Fix 1: drop the units header row (row 0 has "°C (sterility)" etc.)
df = df[df["Crop"] != "—"].reset_index(drop=True)

# Fix 2: rename column
df = df.rename(columns={"Lethal_Drought_Index": "Lethal_Drought_FDI"})

# Fix 3: add season column
# Kharif = Jun-Nov sowing (paddy Kuruvai/Samba, maize, cotton, blackgram etc.)
# Rabi   = Nov-Feb sowing (onion, capsicum, small onion)
# All    = year-round / perennial (banana, coconut, sugarcane, mango etc.)
SEASON_MAP = {
    "paddy":          "Kharif",
    "sugarcane":      "All",
    "banana":         "All",
    "coconut":        "All",
    "tapioca":        "All",
    "mango":          "All",
    "blackgram":      "Kharif",
    "greengram":      "Kharif",
    "gingelly":       "Kharif",
    "brinjal":        "All",
    "cashew":         "All",
    "cotton":         "Kharif",
    "groundnut":      "Kharif",
    "capsicum":       "Rabi",
    "bhendi":         "All",
    "cocoa":          "All",
    "malligai":       "All",
    "guava":          "All",
    "chilli":         "All",
    "annual moringa": "All",
    "snake gourd":    "All",
    "bellary onion":  "Rabi",
    "small onion":    "Rabi",
    "betelvine":      "All",
    "watermelon":     "Kharif",
    "marigold":       "All",
    "tuberose":       "All",
}
df["season"] = df["Crop"].map(SEASON_MAP)
unmapped = df[df["season"].isna()]["Crop"].tolist()
if unmapped:
    print(f"⚠️  Unmapped crops (defaulting to All): {unmapped}")
    df["season"] = df["season"].fillna("All")

# Reorder: Crop, season first, then all limit columns
limit_cols = [c for c in df.columns if c not in ["Crop", "season"]]
df = df[["Crop", "season"] + limit_cols]

df.to_csv("lethal_limits_for_crops.csv", index=False)
print("✅ lethal_limits_for_crops.csv fixed and saved.")
print(f"   Rows: {len(df)}  Columns: {df.columns.tolist()}")
print(df.to_string(index=False))
```

## Section 2 · Fetch Static Geodata (OpenStreetMap)

### Cell 2 · Thanjavur Block Boundaries

```python
# admin_level='6': fetches district-level polygons (admin_level=8 is unavailable
#                  for this area in OSM). Cell 3 filters the result by name to
#                  the 14 required blocks, so this broad fetch is intentional.
# FIX C2-b: ox.features_from_place() returns a MultiIndex GeoDataFrame indexed by
#            (element_type, osmid).  Selecting ['name','geometry'] before
#            reset_index() raises KeyError because those are index levels, not columns.

import osmnx as ox
import geopandas as gpd

print("Fetching Thanjavur taluk boundaries from OpenStreetMap...")

tags = {"admin_level": "6"}       # district-level fetch — admin_level=8 unavailable in OSM for this area;
                                  # Cell 3 filters the output down to the 14 required blocks by name
thanjavur_blocks = ox.features_from_place("Thanjavur, Tamil Nadu, India", tags)

thanjavur_blocks = thanjavur_blocks.reset_index()   # FIX C2-b: flatten MultiIndex first

thanjavur_blocks = thanjavur_blocks[
    thanjavur_blocks.geometry.type.isin(["Polygon", "MultiPolygon"])
].copy()

keep_cols = [c for c in ["name", "geometry"] if c in thanjavur_blocks.columns]
thanjavur_blocks = thanjavur_blocks[keep_cols]

thanjavur_blocks.to_file("thanjavur_blocks.geojson", driver="GeoJSON")

print(f"Fetched {len(thanjavur_blocks)} polygon features.")
print("Names found:", thanjavur_blocks["name"].tolist())
print("Saved → thanjavur_blocks.geojson")
```

### Cell 3 · Filter to the 14 Official Revenue Blocks

```python
# Cell 3 — Filter and map 14 Thanjavur Blocks
import geopandas as gpd
import pandas as pd

print("Generating the 14-block Thanjavur map...")

# 1. Your exact 14 required blocks
THANJAVUR_14_BLOCKS = [
    "Thanjavur", "Kumbakonam", "Papanasam", "Pattukkottai",
    "Peravurani", "Orathanadu", "Thiruvidaimarudur", "Thiruvaiyaru",
    "Budalur", "Ammapettai", "Madukkur", "Sethubhavachatram",
    "Thiruppanandal", "Thirukattupalli"
]

# 2. Map the 6 sub-blocks to the 8 parent geometries found in OSM
parent_mapping = {
    "Thanjavur": "Thanjavur",
    "Kumbakonam": "Kumbakonam",
    "Papanasam": "Papanasam",
    "Pattukkottai": "Pattukkottai",
    "Peravurani": "Peravurani",
    "Orathanadu": "Orathanadu",
    "Thiruvidaimarudur": "Thiruvidaimarudur",
    "Thiruvaiyaru": "Thiruvaiyaru",
    # The sub-blocks linked to their parent taluks:
    "Budalur": "Thiruvaiyaru",           
    "Thirukattupalli": "Thiruvaiyaru",   
    "Ammapettai": "Papanasam",           
    "Thiruppanandal": "Thiruvidaimarudur",
    "Madukkur": "Pattukkottai",          
    "Sethubhavachatram": "Peravurani"    
}

# 3. Load the raw OSM data
gdf = gpd.read_file("thanjavur_blocks.geojson")

# 4. Extract and dissolve the 8 parent shapes
parent_names = list(set(parent_mapping.values()))
parent_gdf = gdf[gdf["name"].str.contains("|".join(parent_names), case=False, na=False)].copy()

def get_parent_name(raw_name):
    for p in parent_names:
        if p.lower() in str(raw_name).lower(): return p
    return raw_name

parent_gdf["parent_name"] = parent_gdf["name"].apply(get_parent_name)
fused_parents = parent_gdf.dissolve(by="parent_name").reset_index()

# 5. Build the final 14-row GeoDataFrame
# NOTE: Sub-blocks (Budalur, Thirukattupalli, Ammapettai, Thiruppanandal,
#       Madukkur, Sethubhavachatram) inherit their parent taluk polygon because
#       OSM does not have separate geometries for them at admin_level=8.
#       They are stored as distinct rows (different "name") but share geometry.
#       Flag this so downstream spatial joins can deduplicate if needed.

final_rows = []
for block in THANJAVUR_14_BLOCKS:
    parent = parent_mapping[block]
    # Grab the geometry of the parent
    geom = fused_parents[fused_parents['parent_name'] == parent].geometry.values[0]
    is_sub = parent_mapping[block] != block  # True if sharing a parent polygon
    final_rows.append({"name": block, "geometry": geom, "shared_geometry": is_sub})

clean_gdf = gpd.GeoDataFrame(final_rows, crs=fused_parents.crs)

# 6. Save it!
clean_gdf.to_file("thanjavur_14_blocks.geojson", driver="GeoJSON")

print(f"✅ Successfully built map with {len(clean_gdf)} blocks!")
for name in sorted(clean_gdf["name"].tolist()):
    print(f"  ✓ {name}")
```

### Cell 4 · Fetch Canal & River Network

```python
# FIX C5-a: blocks_gdf.unary_union is called on a GeoDataFrame — AttributeError.
#            unary_union is a property of GeoSeries, not GeoDataFrame.
#            Fix: blocks_gdf.geometry.unary_union
# FIX C5-b: ox.features_from_polygon also returns a MultiIndex GeoDataFrame.
#            Must reset_index() before accessing 'waterway' or 'name' as columns.
# FIX C5-c: 'waterway' may or may not survive as a column after reset_index()
#            depending on how OSM tags are returned. Guard every column access.

import osmnx as ox
import geopandas as gpd

print("Reading 14-block boundary...")
blocks_gdf = gpd.read_file("thanjavur_14_blocks.geojson")

thanjavur_boundary = blocks_gdf.geometry.unary_union  # unary_union works on all geopandas versions; union_all() requires >= 0.14
print("Fetching canals and rivers from OpenStreetMap… (30–60 sec)")
tags = {"waterway": ["canal", "river"]}
canals_gdf = ox.features_from_polygon(thanjavur_boundary, tags)

canals_gdf = canals_gdf.reset_index()                  # FIX C5-b: flatten MultiIndex

canals_gdf = canals_gdf[
    canals_gdf.geometry.type.isin(["LineString", "MultiLineString", "Polygon"])
].copy()

# FIX C5-c: build cols list only from what actually exists
cols_to_keep = ["geometry"]
for col in ["waterway", "name"]:                        # FIX C5-c: was direct access
    if col in canals_gdf.columns:
        cols_to_keep.append(col)

canals_clean = canals_gdf[cols_to_keep].reset_index(drop=True)
canals_clean.to_file("thanjavur_canals.geojson", driver="GeoJSON")

print(f"Saved {len(canals_clean)} waterway segments → thanjavur_canals.geojson")
if "name" in canals_clean.columns:
    print("Sample waterways:", canals_clean["name"].dropna().unique()[:8].tolist())
```

### Cell 5 · CHIRPS Daily Rainfall 2012–2025 (GEE chunked getInfo)

Fetches all 71,596 block-day records in 28 half-year chunks  
(each ≈ 2,548 features, safely under the 5,000-feature `getInfo()` limit).  
Output: `block_rainfall_2012_2025.csv`

```python
# Cell 5 — CHIRPS block-wise daily rainfall, 2012-2025
# Output : block_rainfall_2012_2025.csv
# Columns: Date | Block | Daily_Rainfall_mm
#
# WHY the previous version was wrong:
#   ee.batch.Export.table.toDrive() silently truncates large collections at
#   GEE's export row cap — producing only one partial chunk (~2,548 rows out
#   of 71,596 needed). It also requires a manual Drive download between runs.
#
# FIX: chunked getInfo() in half-year windows.
#   Each chunk = ~182 days × 14 blocks = ~2,548 features — well within the
#   ~5,000-feature getInfo() hard limit. 28 chunks cover all of 2012-2025.
#   All data lands directly in memory, no Drive export needed.

import ee, json, pandas as pd

ee.Initialize()

with open("thanjavur_14_blocks.geojson", "r") as fh:
    geojson_data = json.load(fh)

thanjavur_fc = ee.FeatureCollection([
    ee.Feature(
        ee.Geometry(feat["geometry"]),
        {"block_name": feat["properties"].get("name", "Unknown")}
    )
    for feat in geojson_data["features"]
])

def extract_block_rain(image):
    date = image.date().format("YYYY-MM-dd")
    return (
        image.select("precipitation")
             .reduceRegions(
                 collection=thanjavur_fc,
                 reducer=ee.Reducer.mean(),
                 scale=5566
             )
             .map(lambda f: f.set("date", date))
    )

# 28 half-year chunks across 2012-2025 — each ~2,548 features, safe for getInfo()
chunks = []
for year in range(2012, 2026):
    chunks.append((f"{year}-01-01", f"{year}-07-01"))
    chunks.append((f"{year}-07-01", f"{year+1}-01-01"))

records = []
for start, end in chunks:
    feats = (
        ee.ImageCollection("UCSB-CHG/CHIRPS/DAILY")
          .filterDate(start, end)
          .map(extract_block_rain)
          .flatten()
          .getInfo()["features"]
    )
    records.extend([f["properties"] for f in feats])
    print(f"  {start} → {end}: {len(feats)} records")

df = pd.DataFrame(records)
df = df.rename(columns={"date": "Date", "block_name": "Block", "mean": "Daily_Rainfall_mm"})
df["Daily_Rainfall_mm"] = pd.to_numeric(df["Daily_Rainfall_mm"], errors="coerce").round(4)

# Sort on datetime object first, format string last — prevents lexicographic mis-sort
df["Date_dt"] = pd.to_datetime(df["Date"])
df = (
    df.drop_duplicates(subset=["Date_dt", "Block"])
      .sort_values(["Block", "Date_dt"])
      .reset_index(drop=True)
)
df["Date"] = df["Date_dt"].dt.strftime("%d-%m-%Y")
df = df[["Date", "Block", "Daily_Rainfall_mm"]]

df.to_csv("block_rainfall_2012_2025.csv", index=False)
print(f"\n✅ Saved → block_rainfall_2012_2025.csv")
print(f"   Rows    : {len(df):,}  (expected 71,596)")
print(f"   Blocks  : {df['Block'].nunique()}")
print(f"   Range   : {df['Date'].iloc[0]} → {df['Date'].iloc[-1]}")
print(df.head(10).to_string())
print("   ...")
print(df.tail(10).to_string())
```

### Cell 6 · SMAP Surface Soil Moisture 2015–2025 (GEE export)

Exports daily 9-km SMAP SPL3SMP_E SSM per block to Google Drive.

```python
# Cell 0d — Fetch NOAA Oceanic Niño Index (ONI) monthly values
# Output: oni_monthly.csv — year | month | ONI
# Required by: SG Model (Module 2) as monthly seasonal feature
#
# Source: NOAA CPC — https://origin.cpc.ncep.noaa.gov/products/analysis_monitoring/
#         ensostuff/ONI_v5.php (official text table)
# Format: fixed-width text, 3-month rolling average of Niño 3.4 SST anomaly

import requests, pandas as pd, io

ONI_URL = (
    "https://origin.cpc.ncep.noaa.gov/products/analysis_monitoring/"
    "ensostuff/ONI_v5.php"
)

# NOAA also publishes a clean text table at this URL:
ONI_TEXT_URL = (
    "https://origin.cpc.ncep.noaa.gov/data/indices/oni.ascii.txt"
)

print("Fetching NOAA ONI index...")
resp = requests.get(ONI_TEXT_URL, timeout=30)
resp.raise_for_status()

# Format: SEAS  YR   TOTAL  CLIMO  ANOM
# SEAS = DJF, JFM, FMA, MAM, AMJ, MJJ, JJA, JAS, ASO, SON, OND, NDJ
# We assign each 3-month season to its middle month.
SEAS_TO_MONTH = {
    "DJF": 1,  "JFM": 2,  "FMA": 3,  "MAM": 4,
    "AMJ": 5,  "MJJ": 6,  "JJA": 7,  "JAS": 8,
    "ASO": 9,  "SON": 10, "OND": 11, "NDJ": 12,
}

lines = resp.text.strip().splitlines()
records = []
for line in lines:
    parts = line.split()
    if len(parts) < 5 or not parts[1].isdigit():
        continue
    seas, yr, total, climo, anom = parts[0], int(parts[1]), parts[2], parts[3], parts[4]
    if seas not in SEAS_TO_MONTH:
        continue
    try:
        oni_val = float(anom)
    except ValueError:
        continue
    records.append({
        "year":  yr,
        "month": SEAS_TO_MONTH[seas],
        "season_code": seas,
        "ONI":   oni_val,
    })

df_oni = pd.DataFrame(records)

# Filter to 2012-2025
df_oni = df_oni[(df_oni["year"] >= 2012) & (df_oni["year"] <= 2025)]
df_oni = df_oni.sort_values(["year", "month"]).reset_index(drop=True)

df_oni.to_csv("oni_monthly.csv", index=False)
print(f"✅ Saved → oni_monthly.csv")
print(f"   Rows: {len(df_oni)}  Range: {df_oni['year'].min()}-{df_oni['year'].max()}")
print(df_oni.head(6).to_string(index=False))
print("   ...")
print(df_oni.tail(6).to_string(index=False))
```

## Section 4 · ERA5-Land Weather 2012–2025

```python
# Cell 7 — ERA5-Land district weather, 2012-2025 — full satellite extraction
# Output : refined_weather_2012-2025.csv
# Columns: Date | Temp_C | T_max_C | T_min_C | Humidity_pct | Wind_Speed_kmh | Solar_Rad_Wm2
#
# - Data pulled entirely from ERA5-Land DAILY_AGGR (no CSV merge, pure satellite)
# - Chunked by year (~365 features/chunk) — within getInfo() limit
# - Sort on datetime object before string formatting → correct daily sequence
# - 4 decimal places on all float columns for high precision

import ee, json, pandas as pd, numpy as np

ee.Initialize()

with open("thanjavur_14_blocks.geojson", "r") as fh:
    geojson_data = json.load(fh)

district_geom = (
    ee.FeatureCollection([
        ee.Feature(ee.Geometry(feat["geometry"]))
        for feat in geojson_data["features"]
    ]).geometry()
)

ERA5_BANDS = [
    "temperature_2m",
    "temperature_2m_max",
    "temperature_2m_min",
    "dewpoint_temperature_2m",
    "surface_solar_radiation_downwards_sum",
    "u_component_of_wind_10m",
    "v_component_of_wind_10m",
]

def extract_weather(img):
    stats = img.reduceRegion(
        reducer=ee.Reducer.mean(),
        geometry=district_geom,
        scale=11132,
        maxPixels=1e9
    )
    return ee.Feature(None, stats).set("date", img.date().format("YYYY-MM-dd"))

records = []
for year in range(2012, 2026):
    start = f"{year}-01-01"
    end   = f"{year+1}-01-01"
    chunk = (
        ee.ImageCollection("ECMWF/ERA5_LAND/DAILY_AGGR")
          .filterDate(start, end)
          .select(ERA5_BANDS)
          .map(extract_weather)
    )
    feats = chunk.getInfo()["features"]
    records.extend([f["properties"] for f in feats])
    print(f"  {year}: {len(feats)} days fetched")

df = pd.DataFrame(records)

# Verify all bands arrived
missing = [b for b in ERA5_BANDS if b not in df.columns]
if missing:
    raise KeyError(f"Missing ERA5 bands: {missing}\nGot: {df.columns.tolist()}")

# ── Unit conversions — 4 decimal places ──────────────────────────────────────
df["Temp_C"]         = (df["temperature_2m"]                        - 273.15).round(4)
df["T_max_C"]        = (df["temperature_2m_max"]                    - 273.15).round(4)
df["T_min_C"]        = (df["temperature_2m_min"]                    - 273.15).round(4)
df["Dew_Point_C"]    = (df["dewpoint_temperature_2m"]               - 273.15).round(4)
df["Solar_Rad_Wm2"]  = (df["surface_solar_radiation_downwards_sum"] / 86400  ).round(4)
df["Wind_Speed_kmh"] = (
    np.sqrt(df["u_component_of_wind_10m"]**2 + df["v_component_of_wind_10m"]**2) * 3.6
).round(4)

# Humidity via Magnus formula on Temp and Dew Point
e_sat = 6.112 * np.exp((17.67 * df["Temp_C"])      / (df["Temp_C"]      + 243.5))
e_act = 6.112 * np.exp((17.67 * df["Dew_Point_C"]) / (df["Dew_Point_C"] + 243.5))
df["Humidity_pct"] = ((e_act / e_sat) * 100).round(4).clip(0, 100)

# ── Sort on datetime FIRST, then format as string ────────────────────────────
# BUG FIX: .sort_values("Date") on a %d-%m-%y string sorts lexicographically
# (day first), not chronologically. This produced the year-jumping order seen
# in the output. Fix: parse to datetime → sort → format last.
df["Date_dt"] = pd.to_datetime(df["date"])   # ISO string from GEE → datetime
df = (
    df.drop_duplicates(subset="Date_dt")
      .sort_values("Date_dt")               # sort on actual datetime, not string
      .reset_index(drop=True)
)
df["Date"] = df["Date_dt"].dt.strftime("%d-%m-%Y")   # 4-digit year, format last

df = df[["Date", "Temp_C", "T_max_C", "T_min_C", "Humidity_pct", "Wind_Speed_kmh", "Solar_Rad_Wm2"]]

df.to_csv("refined_weather_2012-2025.csv", index=False)

print(f"\n✅ Saved → refined_weather_2012-2025.csv")
print(f"   Rows    : {len(df):,}")
print(f"   Columns : {df.columns.tolist()}")
print(f"   Date range: {df['Date'].iloc[0]} → {df['Date'].iloc[-1]}")
print(df.head(10).to_string())
print("   ...")
print(df.tail(10).to_string())
```

```python
# Cell 0b — Soil texture and pH per block via GEE (OpenLandMap)
# Output : soil_texture_blocks.csv
# Columns: block | Sand_fraction | Clay_fraction | SOC_pct | pH | Ksat_mmday
#
# Source  : OpenLandMap hosted on GEE — static, no time dimension.
# Property assets (250m native resolution):
#   Sand : OpenLandMap/SOL/SOL_SAND-WFRACTION_USDA-3A1A1A_M/v02   raw % → ÷100 → [0-1]
#   Clay : OpenLandMap/SOL/SOL_CLAY-WFRACTION_USDA-3A1A1A_M/v02   raw % → ÷100 → [0-1]
#   SOC  : OpenLandMap/SOL/SOL_ORGANIC-CARBON_USDA-6A1C_M/v02     raw g/kg → ÷10 → SOC%
#   pH   : OpenLandMap/SOL/SOL_PH-H2O_USDA-4C1A2A_M/v02           raw pH×10 → ÷10 → pH
#
# Depth: mean of band b0 (0cm) and b10 (10cm) → representative 0-15cm root zone.
#
# Ksat is not available in OpenLandMap.
# Default 5.0 mm/day is used — physically appropriate for Cauvery Delta
# alluvial soils (sandy-loam to clay-loam). Edit the CSV if field data exists.
#
# Column names match exactly what init_soil_block() and evaluate_all_gates() expect:
#   Sand_fraction, Clay_fraction, SOC_pct  → Saxton-Rawls PTF (Cell 9)
#   pH                                     → Gate G6 acid/alkaline check

import ee, json, pandas as pd

ee.Initialize()

with open("thanjavur_14_blocks.geojson", "r") as fh:
    geojson_data = json.load(fh)

thanjavur_fc = ee.FeatureCollection([
    ee.Feature(
        ee.Geometry(feat["geometry"]),
        {"block_name": feat["properties"].get("name", "Unknown")}
    )
    for feat in geojson_data["features"]
])

ASSETS = {
    "Sand_fraction": "OpenLandMap/SOL/SOL_SAND-WFRACTION_USDA-3A1A1A_M/v02",
    "Clay_fraction": "OpenLandMap/SOL/SOL_CLAY-WFRACTION_USDA-3A1A1A_M/v02",
    "SOC_pct":       "OpenLandMap/SOL/SOL_ORGANIC-CARBON_USDA-6A1C_M/v02",
    "pH":            "OpenLandMap/SOL/SOL_PH-H2O_USDA-4C1A2A_M/v02",
}

# Conversion factors: raw GEE value → physical unit
CONVERSIONS = {
    "Sand_fraction": 1 / 100,   # % → fraction [0-1]
    "Clay_fraction": 1 / 100,   # % → fraction [0-1]
    "SOC_pct":       1 / 10,    # g/kg → SOC %
    "pH":            1 / 10,    # pH×10 → pH
}

def fetch_property(col_name, asset_id):
    """
    Averages b0 (0cm) and b10 (10cm) depth bands pixel-wise,
    then takes the spatial mean over each of the 14 block polygons.
    Returns {block_name: converted_float}.
    """
    img       = ee.Image(asset_id).select(["b0", "b10"])
    img_mean  = img.reduce(ee.Reducer.mean())          # pixel-wise mean of two depth bands
    result    = img_mean.reduceRegions(
        collection=thanjavur_fc,
        reducer=ee.Reducer.mean(),                     # spatial mean over block polygon
        scale=250                                      # OpenLandMap native resolution
    ).getInfo()

    factor = CONVERSIONS[col_name]
    out = {}
    for feat in result["features"]:
        p     = feat["properties"]
        block = p.get("block_name")
        raw   = p.get("mean")
        out[block] = round(raw * factor, 6) if raw is not None else None
    return out

print("Fetching OpenLandMap soil properties from GEE...")
results = {}
for col_name, asset_id in ASSETS.items():
    results[col_name] = fetch_property(col_name, asset_id)
    print(f"  ✅ {col_name}")

# Assemble one row per block
blocks = sorted([feat["properties"].get("name") for feat in geojson_data["features"]])
records = []
for block in blocks:
    records.append({
        "block":          block,
        "Sand_fraction":  results["Sand_fraction"].get(block),
        "Clay_fraction":  results["Clay_fraction"].get(block),
        "SOC_pct":        results["SOC_pct"].get(block),
        "pH":             results["pH"].get(block),
        "Ksat_mmday":     5.0,   # default; edit per block if field Ksat data available
    })

df = pd.DataFrame(records)

# Sanity checks
nulls = df.isnull().any(axis=1)
if nulls.any():
    print(f"\n⚠️  {nulls.sum()} block(s) have null values — may be outside OpenLandMap extent:")
    print(df[nulls][["block"]].to_string(index=False))

for col in ["Sand_fraction", "Clay_fraction"]:
    out_of_range = df[(df[col] < 0) | (df[col] > 1)]
    if not out_of_range.empty:
        print(f"⚠️  {col} out of [0,1] range for: {out_of_range['block'].tolist()}")

df.to_csv("soil_texture_blocks.csv", index=False)
print(f"\n✅ Saved → soil_texture_blocks.csv")
print(f"   Rows    : {len(df)}")
print(f"   Columns : {df.columns.tolist()}")
print(df.to_string(index=False))
```

### Cell 9 · Saxton-Rawls PTF — Soil Hydraulic Properties

```python
## Cell 9 · Pre-Season Soil Initialisation (Saxton-Rawls PTF)
# Derives hydraulic properties from SoilGrids data using Saxton-Rawls (2006).
# Outputs per block: theta_FC, theta_WP, FC_mm, WP_mm, TAW, RAW, Ksat, CN, S_NRCS
# Run once per season per block before the daily loop.

import pandas as pd, numpy as np

def saxton_rawls(S, C, OM):
    """
    S  = sand fraction [0-1]
    C  = clay fraction [0-1]
    OM = organic matter fraction [0-1] (SOC × 1.724)
    Returns dict of hydraulic properties.
    """
    theta_FC = (0.299 - 0.251*S + 0.195*C + 0.011*OM
                + 0.006*(S*OM) - 0.027*(C*OM) + 0.452*(S*C))
    theta_WP = (0.031 - 0.024*S + 0.487*C + 0.006*OM
                + 0.005*(S*OM) - 0.013*(C*OM) + 0.068*(S*C))
    # HSG & CN assignment (USDA-NRCS)
    if C > 0.40:
        CN = 89          # HSG-D
    elif C > 0.27:
        CN = 86          # HSG-C
    else:
        CN = 78          # HSG-B (default for alluvial)
    S_NRCS = (25400 / CN) - 254
    return {"theta_FC": theta_FC, "theta_WP": theta_WP,
            "CN": CN, "S_NRCS": S_NRCS}


def init_soil_block(soil_row, crop_stage_row):
    """
    soil_row       : one row from cleaned_thanjavur_full_dataset.csv
    crop_stage_row : current stage row from scientific_crop_stages_fao56.csv
    Returns soil state dict for one block × crop combination.
    """
    S  = soil_row["Sand_fraction"]
    C  = soil_row["Clay_fraction"]
    OM = soil_row["SOC_pct"] * 1.724 / 100   # SOC% → OM fraction
    Zr = crop_stage_row["Zr_stage_m"]
    p  = crop_stage_row["p_fraction"]
    Ksat = soil_row.get("Ksat_mmday", 5.0)   # from SoilGrids ksatc, default 5 mm/day

    props   = saxton_rawls(S, C, OM)
    FC_mm   = props["theta_FC"] * Zr * 1000
    WP_mm   = props["theta_WP"] * Zr * 1000
    TAW     = (props["theta_FC"] - props["theta_WP"]) * Zr * 1000
    RAW     = p * TAW

    return {
        "theta_FC": props["theta_FC"],
        "theta_WP": props["theta_WP"],
        "FC_mm":    FC_mm,
        "WP_mm":    WP_mm,
        "TAW":      TAW,
        "RAW":      RAW,
        "Ksat":     min(Ksat, 50.0),  # cap at 50 mm/day
        "CN":       props["CN"],
        "S_NRCS":   props["S_NRCS"],
    }


print("✅ Saxton-Rawls PTF functions defined.")
print("   Call init_soil_block(soil_row, crop_stage_row) at season start.")
```

### Cell 10 · Reference Evapotranspiration (FAO-56 Penman-Monteith)

```python
## Cell 10 · Reference Evapotranspiration — FAO-56 Penman-Monteith

# FAO-56 Penman-Monteith ETo (Allen et al. 1998, Eq. 6)
# Inputs from ERA5-Land: Temp_C, Humidity_pct, Wind_Speed_kmh, Solar_Rad_Wm2
# u10 → u2 via FAO-56 Eq. 47: u2 = u10 × 4.87 / ln(67.8×10 - 5.42)

import numpy as np

U10_TO_U2 = 4.87 / np.log(67.8 * 10 - 5.42)   # constant wind height correction

def calc_eto(T_mean_C, T_max_C, T_min_C, RH_pct, u10_kmh, Rs_Wm2,
             elevation_m=50.0):
    """
    FAO-56 Penman-Monteith ETo.
    All inputs are daily scalars or numpy arrays.
    Returns ETo in mm/day.

    T_mean_C   : mean daily air temperature (°C)
    T_max_C    : daily max temperature (°C)
    T_min_C    : daily min temperature (°C)
    RH_pct     : mean relative humidity (%)
    u10_kmh    : wind speed at 10m height (km/h)
    Rs_Wm2     : incoming solar radiation (W/m²) — converted to MJ/m²/day
    elevation_m: block elevation above sea level (m), default 50m for delta
    """
    T     = T_mean_C
    Rs    = Rs_Wm2 * 0.0864                        # W/m² → MJ/m²/day
    u2    = (u10_kmh / 3.6) * U10_TO_U2            # km/h → m/s → 2m height

    # Saturation vapour pressure (kPa)
    es    = 0.6108 * np.exp(17.27 * T / (T + 237.3))
    # Actual vapour pressure from RH
    ea    = es * (RH_pct / 100.0)
    # Slope of vapour pressure curve (kPa/°C)
    Delta = 4098 * es / (T + 237.3) ** 2
    # Atmospheric pressure (kPa) from elevation
    P_atm = 101.3 * ((293 - 0.0065 * elevation_m) / 293) ** 5.26
    # Psychrometric constant (kPa/°C)
    gamma = 0.000665 * P_atm

    # Net shortwave radiation (Rns)
    alpha = 0.23                                    # FAO-56 reference crop albedo
    Rns   = (1 - alpha) * Rs
    # Net longwave radiation (Rnl) — simplified Stefan-Boltzmann
    T_max_K = T_max_C + 273.16
    T_min_K = T_min_C + 273.16
    Rnl = (4.903e-9 * ((T_max_K**4 + T_min_K**4) / 2)
           * (0.34 - 0.14 * np.sqrt(ea))
           * (1.35 * np.clip(Rs / (0.75 * Rs + 0.1), 0.3, 1.0) - 0.35))
    Rn  = Rns - Rnl                                # net radiation (MJ/m²/day)
    G   = 0.0                                      # soil heat flux = 0 (daily)

    # FAO-56 PM equation
    numerator   = 0.408 * Delta * (Rn - G) + gamma * (900 / (T + 273)) * u2 * (es - ea)
    denominator = Delta + gamma * (1 + 0.34 * u2)
    ETo = numerator / denominator
    return np.maximum(0.0, ETo)                    # ETo cannot be negative


print("✅ FAO-56 Penman-Monteith ETo function defined.")
print("   calc_eto(T_mean, T_max, T_min, RH, u10_kmh, Rs_Wm2) → mm/day")
```

```python
## Cell 10b · SMAP-Based Groundwater Initialisation
# Replaces the GRACE-FO dependency entirely.
#
# Rationale:
#   - GRACE resolution (~300 km) was never appropriate for 14 small blocks.
#   - SMAP 9-km SSM at sowing time is a well-validated proxy for shallow
#     aquifer state in the Cauvery Delta (thin vadose zone, 1-3 m water table).
#   - A 13-year CHIRPS water-budget spin-up from 2012 → sowing year accumulates
#     cumulative deep percolation (DP), giving a physically consistent GW0.
#
# Outputs per block × sowing date:
#   M0_mm   — initial root-zone moisture from SMAP (mm)
#   GW0_mm  — initial GW storage from spin-up DP accumulation (mm)
#   D0_m    — initial water table depth derived from GW0 and specific yield

import pandas as pd, numpy as np

BLOCKS_14 = [
    "Thanjavur","Kumbakonam","Papanasam","Pattukkottai",
    "Peravurani","Orathanadu","Thiruvidaimarudur","Thiruvaiyaru",
    "Budalur","Ammapettai","Madukkur","Sethubhavachatram",
    "Thirukattupalli","Thiruppanandal"
]

# Default aquifer parameters for shallow alluvial aquifer (Cauvery Delta)
Sy_default   = 0.12   # specific yield
D_surface_m  = 1.5    # assumed saturated thickness at ground surface (m)
alpha_default= 0.004  # GW recession coefficient (day⁻¹)


def init_from_smap(block, sowing_date, smap_df, soil, rain_df, wx_df,
                   crop_stages_df, Sy=Sy_default, spin_start="2012-01-01"):
    """
    Derive M0_mm, GW0_mm, D0_m for one block × sowing_date.

    Parameters
    ----------
    block        : str  — canonical block name
    sowing_date  : str or pd.Timestamp
    smap_df      : DataFrame with columns [Date, block, SSM_m3m3]
    soil         : dict from init_soil_block()
    rain_df      : DataFrame [Date, Block, Daily_Rainfall_mm]
    wx_df        : DataFrame [Date, Temp_C, Humidity_pct, Wind_Speed_kmh, Solar_Rad_Wm2, ...]
    crop_stages_df : DataFrame for a representative "fallow/rice" crop for spin-up
    Sy           : specific yield
    spin_start   : start of water-budget spin-up (default 2012-01-01)

    Returns dict: M0_mm, GW0_mm, D0_m, smap_ssm_raw
    """
    sowing_date = pd.Timestamp(sowing_date)

    # ── Step 1: M0 from SMAP SSM on sowing date (nearest available) ──────────
    block_col = "block" if "block" in smap_df.columns else "Block"
    b_smap = smap_df[smap_df[block_col] == block].copy()
    b_smap["Date_dt"] = pd.to_datetime(b_smap["Date"])
    b_smap = b_smap.sort_values("Date_dt")

    # Find nearest SMAP observation within ±15 days of sowing
    window = b_smap[
        (b_smap["Date_dt"] >= sowing_date - pd.Timedelta(days=15)) &
        (b_smap["Date_dt"] <= sowing_date + pd.Timedelta(days=15))
    ]
    if window.empty:
        # Fallback: use seasonal mean from the same calendar month
        window = b_smap[b_smap["Date_dt"].dt.month == sowing_date.month]

    if window.empty:
        ssm_raw = 0.25   # hard fallback: assume 25% volumetric SMC
        print(f"  ⚠️  [{block}] No SMAP data near {sowing_date.date()} — using fallback 0.25 m³/m³")
    else:
        ssm_raw = float(window.sort_values(
            (window["Date_dt"] - sowing_date).abs()
        ).iloc[0]["SSM_m3m3"])

    # Convert SSM (volumetric, top 5 cm) → root-zone moisture mm
    # SMAP SSM reflects 0-5 cm; scale to root-zone depth via empirical factor
    # (standard approach: multiply by Zr × scaling_factor).
    # We use theta_FC as upper bound and theta_WP as lower bound.
    Zr_init = 0.2   # initial root depth at sowing (m)
    ssm_clamped = float(np.clip(ssm_raw, soil["theta_WP"], soil["theta_FC"]))
    M0_mm = ssm_clamped * Zr_init * 1000

    # ── Step 2: GW0 from CHIRPS water-budget spin-up ─────────────────────────
    # Run a simplified daily water balance from spin_start → sowing_date
    # using CHIRPS rainfall + ERA5 ETo. Accumulate deep percolation (DP_sum).
    b_rain = rain_df[rain_df["Block"] == block].copy()
    b_rain["Date_dt"] = pd.to_datetime(b_rain["Date"], dayfirst=True)

    # Merge rainfall and weather on date
    wx_sub = wx_df.copy()
    wx_sub["Date_dt"] = pd.to_datetime(wx_df["Date"])

    spin_df = pd.merge(
        b_rain[["Date_dt","Daily_Rainfall_mm"]],
        wx_sub[["Date_dt","Temp_C","Humidity_pct","Wind_Speed_kmh","Solar_Rad_Wm2",
                "T_max_C","T_min_C"]].rename(columns={"T_max_C":"T_max","T_min_C":"T_min"}),
        on="Date_dt", how="inner"
    )
    spin_df = spin_df[
        (spin_df["Date_dt"] >= pd.Timestamp(spin_start)) &
        (spin_df["Date_dt"] <  sowing_date)
    ].sort_values("Date_dt").reset_index(drop=True)

    S_NRCS = soil["S_NRCS"]
    Ksat   = soil["Ksat"]
    FC_mm  = soil["FC_mm"]
    WP_mm  = soil["WP_mm"]
    Mt_spin = FC_mm * 0.7   # start spin-up at 70% FC
    DP_sum  = 0.0

    for _, r in spin_df.iterrows():
        Rain  = max(0.0, r["Daily_Rainfall_mm"])
        T_mean = r["Temp_C"]
        T_max  = r.get("T_max", T_mean + 3.0)
        T_min  = r.get("T_min", T_mean - 3.0)
        RH     = r["Humidity_pct"]
        u10    = r["Wind_Speed_kmh"]
        Rs     = r["Solar_Rad_Wm2"]

        if Rain > 0.2 * S_NRCS:
            RO = (Rain - 0.2 * S_NRCS) ** 2 / (Rain + 0.8 * S_NRCS)
        else:
            RO = 0.0
        PIA = Rain - RO

        ETo   = calc_eto(T_mean, T_max, T_min, RH, u10, Rs)
        ETc   = 1.05 * ETo   # assume Kc≈1.05 (fallow/rice surrogate)
        Dr    = max(0.0, FC_mm - Mt_spin)
        RAW_spin = 0.55 * (FC_mm - WP_mm)
        Ks    = 1.0 if Dr <= RAW_spin else max(0.0, (FC_mm-WP_mm-Dr) / ((FC_mm-WP_mm) - RAW_spin))
        ETc_adj = Ks * ETc

        DP    = max(0.0, (Mt_spin + PIA - ETc_adj) - FC_mm)
        Mt_spin = float(np.clip(Mt_spin + PIA - ETc_adj - DP, WP_mm, FC_mm))
        # Apply recession: GW drains each day at rate alpha
        DP_sum = max(0.0, DP_sum * (1 - alpha_default) + DP)

    GW0_mm = DP_sum
    D0_m   = max(0.5, D_surface_m - GW0_mm / (1000 * Sy))

    return {
        "block":        block,
        "sowing_date":  str(sowing_date.date()),
        "smap_ssm_raw": round(ssm_raw, 4),
        "M0_mm":        round(M0_mm, 2),
        "GW0_mm":       round(GW0_mm, 2),
        "D0_m":         round(D0_m, 3),
    }


print("✅ init_from_smap() defined — replaces GRACE-FO dependency.")
print("   Initialises M0 from SMAP SSM at sowing date.")
print("   Initialises GW0 from 13-year CHIRPS water-budget spin-up.")
print("   Call: init_from_smap(block, sowing_date, smap_df, soil, rain_df, wx_df, crop_stages_df)")
```

### Cell 11 · Daily FAO-56 Water Balance Loop (A1–A7)

```python
## Cell 11 · Daily FAO-56 Water Balance Loop (A1–A7)
    
# Full daily hydrological loop per spec formulas A1-A7.
# Runs per block × per crop for the full season.
# Outputs: daily DataFrame with Mt, Ks_t, GWt, Dt, DPt, CRt, ETc_adj

import pandas as pd, numpy as np


def run_water_balance(
    weather_df,        # DataFrame: Date, Temp_C, Humidity_pct, Wind_Speed_kmh,
                       #            Solar_Rad_Wm2, T_max_C, T_min_C, Rainfall_mm
    crop_stages_df,    # DataFrame: filtered rows from scientific_crop_stages_fao56.csv
                       #            for one crop, sorted by GDD_start
    soil,              # dict from init_soil_block()
    M0_mm,             # initial root zone moisture (mm) — from SMAP
    GW0_mm,            # initial GW storage (mm) — from GRACE-FO
    D0_m,              # initial water table depth (m) — from GRACE proxy
    alpha,             # GW recession coefficient (day⁻¹), 0.002–0.008
    Sy=0.12,           # specific yield (alluvial aquifer default)
    sowing_date=None,  # pd.Timestamp — GDD accumulation start
):
    """
    Returns daily_df with columns:
    Date, GDD_daily, GDD_cum, Stage, Stage_Critical,
    Kc, Zr, p, TAW, RAW, FC_mm, WP_mm,
    ETo, ETc, Ks, ETc_adj,
    Rainfall_mm, RO, PIA, CR, DP, Mt, Dr,
    GWt, Dt
    """
    if sowing_date is None:
        sowing_date = weather_df["Date"].iloc[0]

    Tbase   = crop_stages_df["Tbase_C"].iloc[0]
    Ksat    = soil["Ksat"]
    S_NRCS  = soil["S_NRCS"]

    records  = []
    GDD_cum  = 0.0
    Mt       = np.clip(M0_mm, soil["WP_mm"], soil["FC_mm"])
    GWt      = max(0.0, GW0_mm)
    Dt       = D0_m

    for _, row in weather_df[weather_df["Date"] >= sowing_date].iterrows():
        T_mean = row["Temp_C"]
        T_max  = row.get("T_max_C", T_mean + 3.0)   # fallback if not in ERA5 daily
        T_min  = row.get("T_min_C", T_mean - 3.0)
        RH     = row["Humidity_pct"]
        u10    = row["Wind_Speed_kmh"]
        Rs     = row["Solar_Rad_Wm2"]
        Rain   = max(0.0, row.get("Rainfall_mm", 0.0))

        # ── A: GDD & stage lookup ────────────────────────────────────────────
        GDD_daily = max(0.0, (T_max + T_min) / 2.0 - Tbase)
        GDD_cum  += GDD_daily

        stage_row = crop_stages_df[
            (crop_stages_df["GDD_start"] <= GDD_cum) &
            (crop_stages_df["GDD_end"]   >  GDD_cum)
        ]
        if stage_row.empty:                         # past final stage
            stage_row = crop_stages_df.iloc[[-1]]
        stage_row = stage_row.iloc[0]

        Kc       = stage_row["Kc"]
        Zr       = stage_row["Zr_stage_m"]
        p        = stage_row["p_fraction"]
        stage    = stage_row["stage_name"]
        critical = int(stage_row["Stage_Critical"])

        FC_mm  = soil["theta_FC"] * Zr * 1000
        WP_mm  = soil["theta_WP"] * Zr * 1000
        TAW    = (soil["theta_FC"] - soil["theta_WP"]) * Zr * 1000
        RAW    = p * TAW

        # ── A1: Surface runoff (USDA-NRCS CN) ───────────────────────────────
        if Rain > 0.2 * S_NRCS:
            RO  = (Rain - 0.2 * S_NRCS) ** 2 / (Rain + 0.8 * S_NRCS)
        else:
            RO  = 0.0
        PIA = Rain - RO                             # net infiltrated rainfall

        # ── A2: ETo (FAO-56 Penman-Monteith) ────────────────────────────────
        ETo = calc_eto(T_mean, T_max, T_min, RH, u10, Rs)

        # ── A3: ETc and stress coefficient Ks ───────────────────────────────
        ETc = Kc * ETo
        Dr  = FC_mm - Mt                            # depletion from FC
        Dr  = max(0.0, Dr)
        if Dr <= RAW:
            Ks = 1.0
        else:
            Ks = (TAW - Dr) / (TAW - RAW) if (TAW - RAW) > 0 else 0.0
        Ks      = float(np.clip(Ks, 0.0, 1.0))
        ETc_adj = Ks * ETc

        # ── A4: Capillary rise (extinction depth method) ─────────────────────
        if Dt >= 3.0:
            CR = 0.0
        elif Dt <= 0.5:
            CR = min(Ksat * 0.5, 2.0)
        else:
            CR = min(Ksat * 0.5, 2.0) * (1 - (Dt - 0.5) / 2.5)

        # ── A5: Deep percolation ─────────────────────────────────────────────
        DP = max(0.0, (Mt + PIA + CR - ETc_adj) - FC_mm)

        # ── A6: Root zone moisture update ────────────────────────────────────
        Mt = Mt + PIA + CR - ETc_adj - DP
        Mt = float(np.clip(Mt, WP_mm, FC_mm))

        # ── A7: Groundwater storage update ───────────────────────────────────
        Egw = ETo * max(0.0, 1 - (Dt - 0.5) / 2.5) if Dt < 3.0 else 0.0
        GW_prev = GWt
        GWt  = GWt + DP - alpha * GWt - Egw
        GWt  = max(0.0, GWt)
        dGW  = GWt - GW_prev
        Dt   = Dt - dGW / (1000 * Sy)              # update water table depth
        Dt   = max(0.0, Dt)

        records.append({
            "Date":           row["Date"],
            "GDD_daily":      round(GDD_daily, 2),
            "GDD_cum":        round(GDD_cum, 1),
            "Stage":          stage,
            "Stage_Critical": critical,
            "Kc":             round(Kc, 3),
            "Zr_m":           round(Zr, 2),
            "p":              round(p, 3),
            "TAW":            round(TAW, 2),
            "RAW":            round(RAW, 2),
            "FC_mm":          round(FC_mm, 2),
            "WP_mm":          round(WP_mm, 2),
            "ETo":            round(ETo, 3),
            "ETc":            round(ETc, 3),
            "Ks":             round(Ks, 4),
            "ETc_adj":        round(ETc_adj, 3),
            "Rainfall_mm":    round(Rain, 2),
            "RO":             round(RO, 2),
            "PIA":            round(PIA, 2),
            "CR":             round(CR, 3),
            "DP":             round(DP, 3),
            "Mt":             round(Mt, 2),
            "Dr":             round(Dr, 2),
            "GWt":            round(GWt, 2),
            "Dt":             round(Dt, 3),
        })

    return pd.DataFrame(records)


print("✅ run_water_balance() defined.")
print("   Runs formulas A1-A7 for one block × crop over a full season.")
```

### Cell 12 · Season-End GW Classification & WSI

```python
# Section 3.4 — GW class derivation and WSI computation.
# GW_class_projected feeds directly into Gate G3.
#
# GRACE-FO has been dropped. WSI is now derived entirely from:
#   - season-mean GW storage from the water-balance (GWt)
#   - SMAP soil-moisture deficit at sowing (smap_deficit_frac)
#   - rainfall anomaly vs. 2012-2025 block mean (rain_anomaly_neg)

import numpy as np


def classify_gw(daily_df, smap_deficit_frac, rain_anomaly_neg=0.0):
    """
    daily_df           : output of run_water_balance()
    smap_deficit_frac  : normalised SMAP SMC deficit at sowing [0-1]
                         (0 = full FC, 1 = at wilting point)
    rain_anomaly_neg   : normalised negative rainfall anomaly [0-1]
                         (0 = normal rainfall, 1 = worst drought year on record)
                         Compute as: max(0, (mean_rain - season_rain) / mean_rain)

    Returns dict with GW_class_raw, WSI, GW_class_projected
    """
    GW_mean = daily_df["GWt"].mean()

    # Raw class from season-mean GW storage
    if   GW_mean <  15:  GW_class_raw = 1   # CRITICAL
    elif GW_mean <  60:  GW_class_raw = 2   # LOW
    elif GW_mean < 150:  GW_class_raw = 3   # MODERATE
    else:                GW_class_raw = 4   # ADEQUATE

    # Water Stress Index — weighted combination of SMAP deficit + rainfall anomaly
    # (GRACE component removed; SMAP weight increased to compensate)
    WSI = float(np.clip(0.6 * smap_deficit_frac + 0.4 * rain_anomaly_neg, 0.0, 1.0))

    # Projected class — downgrade if WSI > 0.65
    if WSI > 0.65:
        GW_class_projected = max(1, GW_class_raw - int(np.floor(WSI * 2)))
    else:
        GW_class_projected = GW_class_raw

    labels = {1: "CRITICAL", 2: "LOW", 3: "MODERATE", 4: "ADEQUATE"}
    return {
        "GW_mean_mm":          round(GW_mean, 2),
        "GW_class_raw":        GW_class_raw,
        "GW_class_raw_label":  labels[GW_class_raw],
        "WSI":                 round(WSI, 4),
        "GW_class_projected":  GW_class_projected,
        "GW_class_proj_label": labels[GW_class_projected],
    }


print("✅ classify_gw() updated — GRACE term replaced by rain_anomaly_neg.")
print("   Signature: classify_gw(daily_df, smap_deficit_frac, rain_anomaly_neg=0.0)")
print("   Returns GW_class_projected and WSI for Gate G3.")
```

### Cell 13 · Run Module 1 Over All 14 Blocks × Seasons

This cell actually **executes** the water balance — it was the missing  
orchestration layer. It loops over every block and every season year,  
calls `init_from_smap` + `run_water_balance` + `classify_gw`,  
and writes results to `module1_season_results.csv`.

```python
# Cell 13 — Module 1 orchestration: run water balance for all blocks × seasons
# Output: module1_season_results.csv
# Columns: block | season_year | crop | sowing_date | harvest_date |
#           M0_mm | GW0_mm | D0_m | GW_class_projected | WSI |
#           season_ETc_adj_mm | season_rain_mm | season_DP_mm | Ks_mean

import pandas as pd, numpy as np, os

# ── Load required data ────────────────────────────────────────────────────────
rain_df    = pd.read_csv("block_rainfall_2012_2025.csv")
smap_df    = pd.read_csv("smap_ssm_blocks.csv")
wx_df      = pd.read_csv("refined_weather_2012-2025.csv")
soil_df    = pd.read_csv("cleaned_thanjavur_full_dataset.csv")
texture_df = pd.read_csv("soil_texture_blocks.csv")
stages_df  = pd.read_csv("scientific_crop_stages_fao56.csv")   # assumed present

rain_df["Date_dt"]  = pd.to_datetime(rain_df["Date"], dayfirst=True)
wx_df["Date"]       = pd.to_datetime(wx_df["Date"])
wx_df["Date_dt"]    = wx_df["Date"]

# Add rainfall to wx_df if not already there (merge on date, district-level fallback)
if "Rainfall_mm" not in wx_df.columns:
    # Use district-mean rainfall as fallback for ETo denominator
    rain_dist = (rain_df.groupby("Date_dt")["Daily_Rainfall_mm"]
                        .mean().reset_index()
                        .rename(columns={"Daily_Rainfall_mm":"Rainfall_mm"}))
    wx_df = wx_df.merge(rain_dist, on="Date_dt", how="left")
    wx_df["Rainfall_mm"] = wx_df["Rainfall_mm"].fillna(0.0)

BLOCKS_14 = [
    "Thanjavur","Kumbakonam","Papanasam","Pattukkottai",
    "Peravurani","Orathanadu","Thiruvidaimarudur","Thiruvaiyaru",
    "Budalur","Ammapettai","Madukkur","Sethubhavachatram",
    "Thirukattupalli","Thiruppanandal"
]

# Kharif (Samba) paddy sowing: ~Aug 1; season length ~145 days
SEASONS = {yr: {"crop":"paddy","sowing":f"{yr}-08-01","harvest":f"{yr}-12-25"}
           for yr in range(2015, 2025)}   # SMAP available from 2015

# Compute 2012-2025 block-mean seasonal rainfall for anomaly normalisation
rain_season = (rain_df[rain_df["Date_dt"].dt.month.isin(range(6,13))]
               .groupby("Block")["Daily_Rainfall_mm"].mean()
               .rename("mean_daily_rain"))

results = []

for block in BLOCKS_14:
    soil_row = soil_df[soil_df["Blocks"] == block]
    if soil_row.empty:
        print(f"  ⚠️  No soil data for {block} — skipping")
        continue
    soil_row = soil_row.iloc[0]

    tex_row = texture_df[texture_df["block"] == block]
    if tex_row.empty:
        tex_row = texture_df.iloc[0]   # fallback: use first available
    else:
        tex_row = tex_row.iloc[0]

    # Initial root-zone depth for soil init (sowing stage)
    crop_stages_paddy = stages_df[stages_df["crop"].str.lower() == "paddy"].copy()
    if crop_stages_paddy.empty:
        print("  ⚠️  No paddy stages in scientific_crop_stages_fao56.csv — check filename/crop column")
        break
    init_stage_row = crop_stages_paddy.sort_values("GDD_start").iloc[0]
    soil = init_soil_block(soil_row, init_stage_row)

    # Block rainfall for this block
    b_rain = rain_df[rain_df["Block"] == block].copy()

    # Build block-specific weather: merge ERA5 weather with block CHIRPS rainfall
    wx_block = wx_df.copy()
    wx_block = wx_block.merge(
        b_rain[["Date_dt","Daily_Rainfall_mm"]],
        on="Date_dt", how="left"
    )
    # Use block CHIRPS rainfall in preference to ERA5 precip
    if "Daily_Rainfall_mm" in wx_block.columns:
        wx_block["Rainfall_mm"] = wx_block["Daily_Rainfall_mm"].fillna(wx_block["Rainfall_mm"])
    wx_block = wx_block.rename(columns={"Date_dt":"Date"})

    mean_rain = rain_season.get(block, rain_season.mean())

    for yr, season in SEASONS.items():
        crop       = season["crop"]
        sowing     = season["sowing"]
        harvest    = season["harvest"]

        try:
            init = init_from_smap(
                block, sowing, smap_df, soil, b_rain, wx_df,
                crop_stages_paddy
            )
        except Exception as e:
            print(f"  ⚠️  init_from_smap failed [{block} {yr}]: {e}")
            continue

        M0_mm  = init["M0_mm"]
        GW0_mm = init["GW0_mm"]
        D0_m   = init["D0_m"]

        # Slice weather for this season
        wx_season = wx_block[
            (wx_block["Date"] >= pd.Timestamp(sowing)) &
            (wx_block["Date"] <= pd.Timestamp(harvest))
        ].copy()

        if len(wx_season) < 30:
            print(f"  ⚠️  Insufficient weather rows [{block} {yr}]: {len(wx_season)} rows")
            continue

        try:
            daily_df = run_water_balance(
                weather_df     = wx_season,
                crop_stages_df = crop_stages_paddy,
                soil           = soil,
                M0_mm          = M0_mm,
                GW0_mm         = GW0_mm,
                D0_m           = D0_m,
                alpha          = 0.004,
                Sy             = 0.12,
                sowing_date    = pd.Timestamp(sowing),
            )
        except Exception as e:
            print(f"  ⚠️  run_water_balance failed [{block} {yr}]: {e}")
            continue

        season_rain = float(daily_df["Rainfall_mm"].sum())
        rain_deficit = max(0.0, mean_rain * len(daily_df) - season_rain)
        rain_anomaly_neg = float(np.clip(
            rain_deficit / (mean_rain * len(daily_df) + 1e-9), 0.0, 1.0
        ))

        # SMAP deficit fraction at sowing
        ssm_raw = init["smap_ssm_raw"]
        ssm_fc  = soil["theta_FC"]
        ssm_wp  = soil["theta_WP"]
        smap_deficit_frac = float(np.clip(
            (ssm_fc - ssm_raw) / (ssm_fc - ssm_wp + 1e-9), 0.0, 1.0
        ))

        gw_result = classify_gw(daily_df, smap_deficit_frac, rain_anomaly_neg)

        results.append({
            "block":               block,
            "season_year":         yr,
            "crop":                crop,
            "sowing_date":         sowing,
            "harvest_date":        harvest,
            "M0_mm":               M0_mm,
            "GW0_mm":              GW0_mm,
            "D0_m":                D0_m,
            "smap_ssm_raw":        init["smap_ssm_raw"],
            "smap_deficit_frac":   round(smap_deficit_frac, 4),
            "rain_anomaly_neg":    round(rain_anomaly_neg, 4),
            "season_rain_mm":      round(season_rain, 1),
            "season_ETc_adj_mm":   round(float(daily_df["ETc_adj"].sum()), 1),
            "season_DP_mm":        round(float(daily_df["DP"].sum()), 1),
            "Ks_mean":             round(float(daily_df["Ks"].mean()), 4),
            "GW_mean_mm":          gw_result["GW_mean_mm"],
            "GW_class_projected":  gw_result["GW_class_projected"],
            "GW_class_proj_label": gw_result["GW_class_proj_label"],
            "WSI":                 gw_result["WSI"],
        })

        print(f"  ✓ {block} {yr}: GW={gw_result['GW_class_proj_label']}  WSI={gw_result['WSI']:.3f}  Ks={results[-1]['Ks_mean']:.3f}")

results_df = pd.DataFrame(results)
results_df.to_csv("module1_season_results.csv", index=False)
print(f"\n✅ Saved → module1_season_results.csv")
print(f"   Rows    : {len(results_df)}  (expected {len(BLOCKS_14) * len(SEASONS)})")
print(f"   Columns : {list(results_df.columns)}")
print(results_df.head(6).to_string(index=False))
```

## Section 7 · Elimination Gate Engine (G1–G7)

```python
# FIX C6-a: G4 used water_sim_df['Stage'] == 'Mid' to find the flowering window.
#            'Stage' is a label column and may not be present in the sim output.
#            The spec defines the critical window by Stage_Critical == 1 (from
#            scientific_crop_stages_fao56.csv). Fixed to use the flag.
# FIX C6-b: G4 threshold was Lethal_Flowering_Ks directly. Spec says multiply
#            by 0.95 (5 % buffer against hard-edge elimination artifacts).
# FIX C6-c: G5 FDI was stress_days / total_days — a simple day-count fraction.
#            Spec formula B0-FDI is volumetric:
#              FDI = ΣDr_excess / (TAW_mean × N_days)
#              Dr_excess_t = max(0, Dr_t − RAW_t)
#            A crop depleted 5 mm past RAW each day scores very differently from
#            one depleted 50 mm — the count formula cannot distinguish them.
# FIX C6-d: G5 referenced crop_row['Lethal_Drought_Index'] — column does not exist.
#            Correct column (from lethal_limits spec) is 'Lethal_Drought_FDI'.
#            Gate threshold = Lethal_Drought_FDI × 1.10 per spec.
# FIX C6-e: G6 collapsed pH_min and pH_max into one condition with 'or', losing
#            the ability to log which bound was violated. Split into G6a / G6b.

import pandas as pd
import numpy as np


def evaluate_all_gates(
    crop_row: pd.Series,
    max_temp: float,
    total_rain: float,
    soil_ph: float,
    soil_n: float,
    projected_gw_class: int,
    water_sim_df: pd.DataFrame,
) -> tuple:
    """
    Runs one crop through all 7 CDCIS elimination gates.

    water_sim_df required columns
    ─────────────────────────────
    Stage_Critical  int    1 = flowering / critical window, 0 otherwise
    Ks              float  water stress coefficient [0, 1]
    Dr              float  root-zone depletion (mm)
    RAW             float  readily available water (mm)
    TAW             float  total available water (mm)

    Returns
    ───────
    survived : bool
    reasons  : list[str]
    """
    reasons = []

    # ── G1: Heat Sterility ────────────────────────────────────────────────────
    if max_temp >= crop_row["Lethal_Heat_C"]:
        reasons.append(
            f"G1: Peak temp ({max_temp}°C) ≥ sterility limit ({crop_row['Lethal_Heat_C']}°C)"
        )

    # ── G2: Flood / Waterlogging ──────────────────────────────────────────────
    if total_rain >= crop_row["Lethal_Flood_mm"]:
        reasons.append(
            f"G2: Season rain ({total_rain} mm) ≥ flood limit ({crop_row['Lethal_Flood_mm']} mm)"
        )

    # ── G3: Groundwater Class ─────────────────────────────────────────────────
    if projected_gw_class < crop_row["Lethal_GW_Class"]:
        reasons.append(
            f"G3: GW class ({projected_gw_class}) < crop minimum ({crop_row['Lethal_GW_Class']})"
        )

    # ── G6: Soil pH (split into G6a / G6b) ───────────────────────────────────
    # FIX C6-e: was single 'or' condition — now separate for precise logging
    if soil_ph < crop_row["Lethal_pH_min"]:
        reasons.append(
            f"G6a: pH ({soil_ph}) < crop minimum ({crop_row['Lethal_pH_min']})"
        )
    elif soil_ph > crop_row["Lethal_pH_max"]:
        reasons.append(
            f"G6b: pH ({soil_ph}) > crop maximum ({crop_row['Lethal_pH_max']})"
        )

    # ── G7: Soil Nitrogen ─────────────────────────────────────────────────────
    if soil_n < crop_row["Lethal_N_min_kgha"]:
        reasons.append(
            f"G7: Soil N ({soil_n} kg/ha) < crop minimum ({crop_row['Lethal_N_min_kgha']} kg/ha)"
        )

    # ── Dynamic gates — require water simulation output ───────────────────────
    if water_sim_df is not None and not water_sim_df.empty:

        # G4: Flowering Stage Deficit
        # FIX C6-a: use Stage_Critical flag, not stage name string
        # FIX C6-b: threshold = Lethal_Flowering_Ks × 0.95
        flowering = water_sim_df[water_sim_df["Stage_Critical"] == 1]
        if not flowering.empty:
            mean_ks      = flowering["Ks"].mean()
            ks_threshold = round(crop_row["Lethal_Flowering_Ks"] * 0.95, 6)  # round() eliminates float noise
            if mean_ks < ks_threshold:
                reasons.append(
                    f"G4: Mean Ks during flowering ({mean_ks:.3f}) < "
                    f"threshold ({ks_threshold:.3f}  |  Lethal={crop_row['Lethal_Flowering_Ks']})"
                )

        # G5: Cumulative Drought — Farm Drought Index (spec formula B0-FDI)
        # FIX C6-c: volumetric formula, not day-count fraction
        # FIX C6-d: column is 'Lethal_Drought_FDI', gate = value × 1.10
        sim               = water_sim_df.copy()
        sim["Dr_excess"]  = np.maximum(0.0, sim["Dr"] - sim["RAW"])
        sigma_dr_excess   = sim["Dr_excess"].sum()
        taw_mean          = sim["TAW"].mean()
        n_days            = len(sim)
        fdi               = sigma_dr_excess / (taw_mean * n_days) if taw_mean * n_days > 0 else 0.0
        fdi_gate          = round(crop_row["Lethal_Drought_FDI"] * 1.10, 6)  # FIX C6-d + FIX C6-f: round() eliminates IEEE 754 float noise (0.40×1.10=0.44000...6)
        if fdi >= fdi_gate:
            reasons.append(
                f"G5: FDI ({fdi:.3f}) ≥ gate ({fdi_gate:.3f}  |  "
                f"Lethal={crop_row['Lethal_Drought_FDI']})"
            )

    return len(reasons) == 0, reasons


# ── Smoke test ────────────────────────────────────────────────────────────────
print("Running smoke test…")

dummy_crop = pd.Series({
    "Crop":                 "paddy",
    "Lethal_Heat_C":        38.0,
    "Lethal_Flood_mm":    1800.0,
    "Lethal_GW_Class":       2,
    "Lethal_pH_min":         5.0,
    "Lethal_pH_max":         8.0,
    "Lethal_N_min_kgha":    80.0,
    "Lethal_Flowering_Ks":   0.65,
    "Lethal_Drought_FDI":    0.40,
})

dummy_sim = pd.DataFrame({
    "Stage_Critical": [0]*30 + [1]*60 + [0]*30,
    "Ks":             [1.0]*80 + [0.45]*40,
    "Dr":             [20.0]*80 + [80.0]*40,
    "RAW":            [50.0]*120,
    "TAW":            [100.0]*120,
})

survived, reasons = evaluate_all_gates(
    crop_row=dummy_crop,
    max_temp=36.0, total_rain=1200.0,
    soil_ph=6.5,   soil_n=100.0,
    projected_gw_class=3,
    water_sim_df=dummy_sim,
)

print(f"Survived: {survived}")
for r in reasons:
    print(f"  ✗ {r}")
if not reasons:
    print("  ✓ All gates passed")
print("\n✅ Gate engine ready.")
```
